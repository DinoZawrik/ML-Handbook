    *   **–§–æ—Ä–º—É–ª–∞:** $Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$
    *   **–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:** –î–æ–ª—è –≤—Å–µ—Ö –≤–µ—Ä–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤.
    *   **–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ:** –ú–æ–∂–µ—Ç –±—ã—Ç—å –æ–±–º–∞–Ω—á–∏–≤–æ –≤—ã—Å–æ–∫–æ–π –Ω–∞ **–Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö**.
2.  **Precision (–¢–æ—á–Ω–æ—Å—Ç—å):**
    *   **–§–æ—Ä–º—É–ª–∞:** $Precision = \frac{TP}{TP + FP}$
    *   **–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:** –î–æ–ª—è –æ–±—ä–µ–∫—Ç–æ–≤, –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —è–≤–ª—è—é—â–∏—Ö—Å—è Positive, —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö, —á—Ç–æ –º–æ–¥–µ–ª—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–ª–∞ –∫–∞–∫ Positive. –û—Ç–≤–µ—á–∞–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å: "–ï—Å–ª–∏ –º–æ–¥–µ–ª—å —Å–∫–∞–∑–∞–ª–∞ Positive, –Ω–∞—Å–∫–æ–ª—å–∫–æ –µ–π –≤–µ—Ä–∏—Ç—å?"
    *   **–í–∞–∂–Ω–æ:** –í—ã—Å–æ–∫–∏–π Precision –æ–∑–Ω–∞—á–∞–µ—Ç –º–∞–ª–æ –ª–æ–∂–Ω—ã—Ö —Å—Ä–∞–±–∞—Ç—ã–≤–∞–Ω–∏–π (FP).
3.  **Recall (–ü–æ–ª–Ω–æ—Ç–∞ / –ß—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å / TPR):**
    *   **–§–æ—Ä–º—É–ª–∞:** $Recall = \frac{TP}{TP + FN}$
    *   **–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:** –î–æ–ª—è –æ–±—ä–µ–∫—Ç–æ–≤, –≤–µ—Ä–Ω–æ –Ω–∞–π–¥–µ–Ω–Ω—ã—Ö –º–æ–¥–µ–ª—å—é –∫–∞–∫ Positive, —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö —Ä–µ–∞–ª—å–Ω–æ Positive –æ–±—ä–µ–∫—Ç–æ–≤. –û—Ç–≤–µ—á–∞–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å: "–°–∫–æ–ª—å–∫–æ —Ä–µ–∞–ª—å–Ω—ã—Ö Positive –æ–±—ä–µ–∫—Ç–æ–≤ –º–æ–¥–µ–ª—å –Ω–∞—à–ª–∞?"
    *   **–í–∞–∂–Ω–æ:** –í—ã—Å–æ–∫–∏–π Recall –æ–∑–Ω–∞—á–∞–µ—Ç –º–∞–ª–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ —Ü–µ–ª–∏ (FN).
4.  **F1-score (F-–º–µ—Ä–∞):**
    *   **–§–æ—Ä–º—É–ª–∞:** $F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall} = \frac{2 \times TP}{2 \times TP + FP + FN}$
    *   **–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:** –ì–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ Precision –∏ Recall. –î–∞–µ—Ç –±–∞–ª–∞–Ω—Å –º–µ–∂–¥—É –Ω–∏–º–∏.
    *   **–í–∞–∂–Ω–æ:** –û—Å–æ–±–µ–Ω–Ω–æ –ø–æ–ª–µ–∑–Ω–∞ –¥–ª—è **–Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö**, –∫–æ–≥–¥–∞ –≤–∞–∂–Ω—ã –∏ Precision, –∏ Recall. –í—ã—Å–æ–∫–∏–π F1 –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –µ—Å–ª–∏ –æ–±–µ –º–µ—Ç—Ä–∏–∫–∏ –≤—ã—Å–æ–∫–∏.
5.  **ROC AUC (Area Under the Receiver Operating Characteristic Curve):**
    *   **ROC-–∫—Ä–∏–≤–∞—è:** –ì—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ **TPR (Recall)** –æ—Ç **FPR (False Positive Rate)** –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ –ø–æ—Ä–æ–≥–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏. ($FPR = \frac{FP}{FP + TN}$)
    *   **–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è AUC:** –ü–ª–æ—â–∞–¥—å –ø–æ–¥ ROC-–∫—Ä–∏–≤–æ–π. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –º–æ–¥–µ–ª—å —Ä–∞–Ω–∂–∏—Ä—É–µ—Ç —Å–ª—É—á–∞–π–Ω–æ –≤—ã–±—Ä–∞–Ω–Ω—ã–π Positive –æ–±—ä–µ–∫—Ç –≤—ã—à–µ, —á–µ–º —Å–ª—É—á–∞–π–Ω–æ –≤—ã–±—Ä–∞–Ω–Ω—ã–π Negative.
    *   **–°–≤–æ–π—Å—Ç–≤–∞:**
        *   AUC = 1: –ò–¥–µ–∞–ª—å–Ω—ã–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä.
        *   AUC = 0.5: –°–ª—É—á–∞–π–Ω—ã–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä.
        *   AUC > 0.5: –ú–æ–¥–µ–ª—å –ª—É—á—à–µ —Å–ª—É—á–∞–π–Ω–æ–π.
    *   **–í–∞–∂–Ω–æ:** –û—Ü–µ–Ω–∏–≤–∞–µ—Ç –∫–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–∏ **–Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –ø–æ—Ä–æ–≥–∞**. –û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ —É—Å—Ç–æ–π—á–∏–≤–∞ –∫ –¥–∏—Å–±–∞–ª–∞–Ω—Å—É –∫–ª–∞—Å—Å–æ–≤.
6.  **PR AUC (Area Under the Precision-Recall Curve):**
    *   **PR-–∫—Ä–∏–≤–∞—è:** –ì—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ **Precision** –æ—Ç **Recall** –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ –ø–æ—Ä–æ–≥–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏.
    *   **–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è AUC:** –ü–ª–æ—â–∞–¥—å –ø–æ–¥ PR-–∫—Ä–∏–≤–æ–π. –ê–≥—Ä–µ–≥–∏—Ä—É–µ—Ç Precision –ø–æ —Ä–∞–∑–Ω—ã–º —É—Ä–æ–≤–Ω—è–º Recall.
    *   **–°–≤–æ–π—Å—Ç–≤–∞:**
        *   AUC = 1: –ò–¥–µ–∞–ª—å–Ω—ã–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä (–≤ —Ç–µ—Ä–º–∏–Ω–∞—Ö P –∏ R).
        *   –ë–∞–∑–æ–≤—ã–π —É—Ä–æ–≤–µ–Ω—å (—Å–ª—É—á–∞–π–Ω—ã–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä) –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –¥–æ–ª–∏ Positive –∫–ª–∞—Å—Å–∞ (–Ω–∏–∂–µ 0.5 –ø—Ä–∏ –¥–∏—Å–±–∞–ª–∞–Ω—Å–µ).
    *   **–í–∞–∂–Ω–æ:** –ë–æ–ª–µ–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω–∞, —á–µ–º ROC AUC, –Ω–∞ **—Å–∏–ª—å–Ω–æ –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö**, –∫–æ–≥–¥–∞ –≤–∞–∂–Ω–∞ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∏–º–µ–Ω–Ω–æ –Ω–∞ —Ä–µ–¥–∫–æ–º Positive –∫–ª–∞—Å—Å–µ.

---

### IV.C. –ü–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ, –ù–µ–¥–æ–æ–±—É—á–µ–Ω–∏–µ –∏ –ö–æ–º–ø—Ä–æ–º–∏—Å—Å –°–º–µ—â–µ–Ω–∏—è-–î–∏—Å–ø–µ—Ä—Å–∏–∏

**1. –ù–µ–¥–æ–æ–±—É—á–µ–Ω–∏–µ (Underfitting):**
*   **–ü—Ä–æ–±–ª–µ–º–∞:** –ú–æ–¥–µ–ª—å **—Å–ª–∏—à–∫–æ–º –ø—Ä–æ—Å—Ç–∞**, –Ω–µ –º–æ–∂–µ—Ç —É–ª–æ–≤–∏—Ç—å –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–∏ –≤ –¥–∞–Ω–Ω—ã—Ö.
*   **–•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏:** **–ü–ª–æ—Ö–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ** –Ω–∞ *–æ–±—É—á–∞—é—â–µ–π* –∏ *—Ç–µ—Å—Ç–æ–≤–æ–π* –≤—ã–±–æ—Ä–∫–∞—Ö.
*   **–ü—Ä–∏—á–∏–Ω–∞:** **–í—ã—Å–æ–∫–æ–µ —Å–º–µ—â–µ–Ω–∏–µ (High Bias)** - —Å–∏–ª—å–Ω—ã–µ —É–ø—Ä–æ—â–∞—é—â–∏–µ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è –º–æ–¥–µ–ª–∏. **–ù–∏–∑–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è (Low Variance)** - –º–æ–¥–µ–ª—å –Ω–µ—á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–∞ –∫ –¥–∞–Ω–Ω—ã–º.
*   **–ë–æ—Ä—å–±–∞:** –£—Å–ª–æ–∂–Ω–∏—Ç—å –º–æ–¥–µ–ª—å, –¥–æ–±–∞–≤–∏—Ç—å/—É–ª—É—á—à–∏—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∏, —É–º–µ–Ω—å—à–∏—Ç—å —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—é.

**2. –ü–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ (Overfitting):**
*   **–ü—Ä–æ–±–ª–µ–º–∞:** –ú–æ–¥–µ–ª—å **—Å–ª–∏—à–∫–æ–º —Å–ª–æ–∂–Ω–∞**, "–∑–∞–ø–æ–º–∏–Ω–∞–µ—Ç" –æ–±—É—á–∞—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ (–≤–∫–ª—é—á–∞—è —à—É–º), –Ω–æ –ø–ª–æ—Ö–æ –æ–±–æ–±—â–∞–µ—Ç –Ω–∞ –Ω–æ–≤—ã–µ.
*   **–•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏:** **–û—Ç–ª–∏—á–Ω–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ** –Ω–∞ *–æ–±—É—á–∞—é—â–µ–π* –≤—ã–±–æ—Ä–∫–µ, **–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ —Ö—É–¥—à–µ–µ** –Ω–∞ *—Ç–µ—Å—Ç–æ–≤–æ–π*. –ë–æ–ª—å—à–æ–π —Ä–∞–∑—Ä—ã–≤ –º–µ–∂–¥—É –º–µ—Ç—Ä–∏–∫–∞–º–∏ –Ω–∞ –æ–±—É—á–µ–Ω–∏–∏ –∏ —Ç–µ—Å—Ç–µ (Overfitting Gap).
*   **–ü—Ä–∏—á–∏–Ω–∞:** **–í—ã—Å–æ–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è (High Variance)** - —Å–∏–ª—å–Ω–∞—è —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∫ –æ–±—É—á–∞—é—â–∏–º –¥–∞–Ω–Ω—ã–º. **–ù–∏–∑–∫–æ–µ —Å–º–µ—â–µ–Ω–∏–µ (Low Bias)** - –º–æ–¥–µ–ª—å —Ö–æ—Ä–æ—à–æ –æ–ø–∏—Å—ã–≤–∞–µ—Ç –æ–±—É—á–µ–Ω–∏–µ.
*   **–ë–æ—Ä—å–±–∞:** –£–ø—Ä–æ—Å—Ç–∏—Ç—å –º–æ–¥–µ–ª—å, —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è, –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è, —Ä–∞–Ω–Ω—è—è –æ—Å—Ç–∞–Ω–æ–≤–∫–∞, –æ—Ç–±–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∞–Ω—Å–∞–º–±–ª–∏, –±–æ–ª—å—à–µ –¥–∞–Ω–Ω—ã—Ö.

**3. –ö–æ–º–ø—Ä–æ–º–∏—Å—Å –°–º–µ—â–µ–Ω–∏—è-–î–∏—Å–ø–µ—Ä—Å–∏–∏ (Bias-Variance Tradeoff):**
*   **–°—É—Ç—å:** –ù–µ–ª—å–∑—è –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∏ Bias, –∏ Variance. –£–º–µ–Ω—å—à–µ–Ω–∏–µ –æ–¥–Ω–æ–≥–æ —á–∞—Å—Ç–æ –≤–µ–¥–µ—Ç –∫ —É–≤–µ–ª–∏—á–µ–Ω–∏—é –¥—Ä—É–≥–æ–≥–æ.
*   **–¶–µ–ª—å:** –ù–∞–π—Ç–∏ **–±–∞–ª–∞–Ω—Å** –º–µ–∂–¥—É Bias –∏ Variance –¥–ª—è –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ **–æ–±—â–µ–π –æ—à–∏–±–∫–∏ –Ω–∞ –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö**.
*   **–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è:** U-–æ–±—Ä–∞–∑–Ω–∞—è –∫—Ä–∏–≤–∞—è –æ—à–∏–±–∫–∏ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Å–ª–æ–∂–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏. –ú–∏–Ω–∏–º—É–º –æ–±—â–µ–π –æ—à–∏–±–∫–∏ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è –ø—Ä–∏ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–π —Å–ª–æ–∂–Ω–æ—Å—Ç–∏.

**(–†–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è - –∫–ª—é—á–µ–≤–æ–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç —É–ø—Ä–∞–≤–ª–µ–Ω–∏—è —ç—Ç–∏–º –∫–æ–º–ø—Ä–æ–º–∏—Å—Å–æ–º, —Å–º. —Ä–∞–∑–¥–µ–ª –æ –õ–∏–Ω–µ–π–Ω–æ–π –†–µ–≥—Ä–µ—Å—Å–∏–∏).**

---
–≠—Ç–æ—Ç –≤–∞—Ä–∏–∞–Ω—Ç —É–±–∏—Ä–∞–µ—Ç –ø–æ–≤—Ç–æ—Ä–Ω—ã–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, —Å–æ–∫—Ä–∞—â–∞–µ—Ç –æ–ø–∏—Å–∞–Ω–∏—è, –Ω–æ —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç —Å—É—Ç—å –∫–∞–∂–¥–æ–≥–æ –ø–æ–Ω—è—Ç–∏—è, —Ñ–æ—Ä–º—É–ª—ã –∏ –æ—Å–Ω–æ–≤–Ω—ã–µ –≤—ã–≤–æ–¥—ã. –ö–æ–¥ —Å–≤–µ–¥–µ–Ω –∫ –ø—Ä–∏–º–µ—Ä–∞–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è `sklearn` –∏–ª–∏ —Å—Å—ã–ª–∫–∞–º –Ω–∞ –ø—Ä–µ–¥—ã–¥—É—â–∏–µ –±–ª–æ–∫–∏.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression # –ò—Å–ø–æ–ª—å–∑—É–µ–º –ø—Ä–æ—Å—Ç–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä –¥–ª—è –ø—Ä–∏–º–µ—Ä–∞
from sklearn.metrics import (accuracy_score,
                             precision_score,
                             recall_score,
                             f1_score,
                             confusion_matrix,
                             ConfusionMatrixDisplay,
                             roc_auc_score,
                             roc_curve,             # –î–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è ROC-–∫—Ä–∏–≤–æ–π
                             precision_recall_curve, # –î–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è PR-–∫—Ä–∏–≤–æ–π
                             auc                    # –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø–ª–æ—â–∞–¥–∏ –ø–æ–¥ –∫—Ä–∏–≤–æ–π (PR AUC)
                            )

# 1. –ì–µ–Ω–µ—Ä–∞—Ü–∏—è —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
# –°–¥–µ–ª–∞–µ–º –∫–ª–∞—Å—Å—ã –Ω–µ–º–Ω–æ–≥–æ –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏ –º–µ—Ç—Ä–∏–∫
X, y = make_classification(n_samples=1000, n_features=20, n_informative=10,
                           n_redundant=5, n_classes=2, weights=[0.8, 0.2], # 80% –∫–ª–∞—Å—Å 0, 20% –∫–ª–∞—Å—Å 1
                           flip_y=0.05, random_state=42)

# 2. –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# 3. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ (–õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –†–µ–≥—Ä–µ—Å—Å–∏—è –¥–ª—è –ø—Ä–∏–º–µ—Ä–∞)
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

# 4. –ü–æ–ª—É—á–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
y_pred = model.predict(X_test)              # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∫–ª–∞—Å—Å—ã
y_proba = model.predict_proba(X_test)       # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ [P(class=0), P(class=1)]
y_proba_positive = y_proba[:, 1]            # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Ç–æ–ª—å–∫–æ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞ (–∫–ª–∞—Å—Å–∞ 1)

print("--- –†–∞—Å—á–µ—Ç –∏ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫ ---")

# 5. Confusion Matrix (–ú–∞—Ç—Ä–∏—Ü–∞ –û—à–∏–±–æ–∫)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)

print("\nConfusion Matrix:")
# –í—ã–≤–µ–¥–µ–º —Ç–µ–∫—Å—Ç–æ–≤—ã–π –≤–∞—Ä–∏–∞–Ω—Ç –¥–ª—è —è—Å–Ω–æ—Å—Ç–∏
tn, fp, fn, tp = cm.ravel()
print(f"TN: {tn}, FP: {fp}")
print(f"FN: {fn}, TP: {tp}")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è Confusion Matrix
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
plt.close() # –ó–∞–∫—Ä—ã–≤–∞–µ–º –≥—Ä–∞—Ñ–∏–∫ –º–∞—Ç—Ä–∏—Ü—ã –æ—à–∏–±–æ–∫

# 6. Accuracy, Precision, Recall, F1-score
accuracy = accuracy_score(y_test, y_pred)
# –†–∞—Å—á–µ—Ç –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞ (pos_label=1 –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"\nAccuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")

# 7. ROC AUC Score
roc_auc = roc_auc_score(y_test, y_proba_positive)
print(f"ROC AUC:   {roc_auc:.4f}")

# 8. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ ROC-–∫—Ä–∏–≤–æ–π
fpr, tpr, roc_thresholds = roc_curve(y_test, y_proba_positive)

plt.figure(figsize=(7, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random classifier') # –õ–∏–Ω–∏—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ —É–≥–∞–¥—ã–≤–∞–Ω–∏—è
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
plt.close() # –ó–∞–∫—Ä—ã–≤–∞–µ–º –≥—Ä–∞—Ñ–∏–∫ ROC

# 9. PR AUC Score –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ PR-–∫—Ä–∏–≤–æ–π
precision_curve, recall_curve, pr_thresholds = precision_recall_curve(y_test, y_proba_positive)
pr_auc = auc(recall_curve, precision_curve) # –í—ã—á–∏—Å–ª—è–µ–º –ø–ª–æ—â–∞–¥—å –ø–æ–¥ PR-–∫—Ä–∏–≤–æ–π
print(f"\nPR AUC:    {pr_auc:.4f}")

plt.figure(figsize=(7, 6))
plt.plot(recall_curve, precision_curve, color='blue', lw=2, label=f'PR curve (area = {pr_auc:.2f})')
# –ë–∞–∑–æ–≤—ã–π —É—Ä–æ–≤–µ–Ω—å –¥–ª—è PR AUC (–¥–æ–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞)
baseline = y_test.sum() / len(y_test)
plt.plot([0, 1], [baseline, baseline], linestyle='--', color='red', label=f'Baseline (No Skill = {baseline:.2f})')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower left")
plt.grid(True)
plt.show()
plt.close() # –ó–∞–∫—Ä—ã–≤–∞–µ–º –≥—Ä–∞—Ñ–∏–∫ PR
```

### IV.C. –ü–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ, –ù–µ–¥–æ–æ–±—É—á–µ–Ω–∏–µ –∏ –ö–æ–º–ø—Ä–æ–º–∏—Å—Å –°–º–µ—â–µ–Ω–∏—è-–î–∏—Å–ø–µ—Ä—Å–∏–∏

**1. –ù–µ–¥–æ–æ–±—É—á–µ–Ω–∏–µ (Underfitting):**
*   **–ü—Ä–æ–±–ª–µ–º–∞:** –ú–æ–¥–µ–ª—å **—Å–ª–∏—à–∫–æ–º –ø—Ä–æ—Å—Ç–∞**, –Ω–µ –º–æ–∂–µ—Ç —É–ª–æ–≤–∏—Ç—å –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–∏ –≤ –¥–∞–Ω–Ω—ã—Ö.
*   **–•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏:** **–ü–ª–æ—Ö–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ** –Ω–∞ *–æ–±—É—á–∞—é—â–µ–π* –∏ *—Ç–µ—Å—Ç–æ–≤–æ–π* –≤—ã–±–æ—Ä–∫–∞—Ö.
*   **–ü—Ä–∏—á–∏–Ω–∞:** **–í—ã—Å–æ–∫–æ–µ —Å–º–µ—â–µ–Ω–∏–µ (High Bias)** - —Å–∏–ª—å–Ω—ã–µ —É–ø—Ä–æ—â–∞—é—â–∏–µ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è –º–æ–¥–µ–ª–∏. **–ù–∏–∑–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è (Low Variance)** - –º–æ–¥–µ–ª—å –Ω–µ—á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–∞ –∫ –¥–∞–Ω–Ω—ã–º.
*   **–ë–æ—Ä—å–±–∞:** –£—Å–ª–æ–∂–Ω–∏—Ç—å –º–æ–¥–µ–ª—å, –¥–æ–±–∞–≤–∏—Ç—å/—É–ª—É—á—à–∏—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∏, —É–º–µ–Ω—å—à–∏—Ç—å —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—é.

**2. –ü–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ (Overfitting):**
*   **–ü—Ä–æ–±–ª–µ–º–∞:** –ú–æ–¥–µ–ª—å **—Å–ª–∏—à–∫–æ–º —Å–ª–æ–∂–Ω–∞**, "–∑–∞–ø–æ–º–∏–Ω–∞–µ—Ç" –æ–±—É—á–∞—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ (–≤–∫–ª—é—á–∞—è —à—É–º), –Ω–æ –ø–ª–æ—Ö–æ –æ–±–æ–±—â–∞–µ—Ç –Ω–∞ –Ω–æ–≤—ã–µ.
*   **–•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏:** **–û—Ç–ª–∏—á–Ω–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ** –Ω–∞ *–æ–±—É—á–∞—é—â–µ–π* –≤—ã–±–æ—Ä–∫–µ, **–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ —Ö—É–¥—à–µ–µ** –Ω–∞ *—Ç–µ—Å—Ç–æ–≤–æ–π*. –ë–æ–ª—å—à–æ–π —Ä–∞–∑—Ä—ã–≤ –º–µ–∂–¥—É –º–µ—Ç—Ä–∏–∫–∞–º–∏ –Ω–∞ –æ–±—É—á–µ–Ω–∏–∏ –∏ —Ç–µ—Å—Ç–µ (Overfitting Gap).
*   **–ü—Ä–∏—á–∏–Ω–∞:** **–í—ã—Å–æ–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è (High Variance)** - —Å–∏–ª—å–Ω–∞—è —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∫ –æ–±—É—á–∞—é—â–∏–º –¥–∞–Ω–Ω—ã–º. **–ù–∏–∑–∫–æ–µ —Å–º–µ—â–µ–Ω–∏–µ (Low Bias)** - –º–æ–¥–µ–ª—å —Ö–æ—Ä–æ—à–æ –æ–ø–∏—Å—ã–≤–∞–µ—Ç –æ–±—É—á–µ–Ω–∏–µ.
*   **–ë–æ—Ä—å–±–∞:** –£–ø—Ä–æ—Å—Ç–∏—Ç—å –º–æ–¥–µ–ª—å, —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è, –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è, —Ä–∞–Ω–Ω—è—è –æ—Å—Ç–∞–Ω–æ–≤–∫–∞, –æ—Ç–±–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∞–Ω—Å–∞–º–±–ª–∏, –±–æ–ª—å—à–µ –¥–∞–Ω–Ω—ã—Ö.

**3. –ö–æ–º–ø—Ä–æ–º–∏—Å—Å –°–º–µ—â–µ–Ω–∏—è-–î–∏—Å–ø–µ—Ä—Å–∏–∏ (Bias-Variance Tradeoff):**
*   **–°—É—Ç—å:** –ù–µ–ª—å–∑—è –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∏ Bias, –∏ Variance. –£–º–µ–Ω—å—à–µ–Ω–∏–µ –æ–¥–Ω–æ–≥–æ —á–∞—Å—Ç–æ –≤–µ–¥–µ—Ç –∫ —É–≤–µ–ª–∏—á–µ–Ω–∏—é –¥—Ä—É–≥–æ–≥–æ.
*   **–¶–µ–ª—å:** –ù–∞–π—Ç–∏ **–±–∞–ª–∞–Ω—Å** –º–µ–∂–¥—É Bias –∏ Variance –¥–ª—è –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ **–æ–±—â–µ–π –æ—à–∏–±–∫–∏ –Ω–∞ –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö**.
*   **–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è:** U-–æ–±—Ä–∞–∑–Ω–∞—è –∫—Ä–∏–≤–∞—è –æ—à–∏–±–∫–∏ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Å–ª–æ–∂–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏. –ú–∏–Ω–∏–º—É–º –æ–±—â–µ–π –æ—à–∏–±–∫–∏ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è –ø—Ä–∏ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–π —Å–ª–æ–∂–Ω–æ—Å—Ç–∏.

**(–†–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è - –∫–ª—é—á–µ–≤–æ–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç —É–ø—Ä–∞–≤–ª–µ–Ω–∏—è —ç—Ç–∏–º –∫–æ–º–ø—Ä–æ–º–∏—Å—Å–æ–º, —Å–º. —Ä–∞–∑–¥–µ–ª –æ –õ–∏–Ω–µ–π–Ω–æ–π –†–µ–≥—Ä–µ—Å—Å–∏–∏).**
