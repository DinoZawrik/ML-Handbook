    *   **–ö–∞–∫ —Å—á–∏—Ç–∞–µ—Ç—Å—è:**  –ü–æ—Ö–æ–∂–µ –Ω–∞ MSE, –Ω–æ —Ä–∞–∑–Ω–æ—Å—Ç—å –º–µ–∂–¥—É –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º –∏ —Ä–µ–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –±–µ—Ä–µ—Ç—Å—è –ø–æ *–∞–±—Å–æ–ª—é—Ç–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é* (–º–æ–¥—É–ª—é), –∞ –Ω–µ –≤–æ–∑–≤–æ–¥–∏—Ç—Å—è –≤ –∫–≤–∞–¥—Ä–∞—Ç.  –ó–∞—Ç–µ–º —ç—Ç–∏ –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ –æ—à–∏–±–∫–∏ —É—Å—Ä–µ–¥–Ω—è—é—Ç—Å—è.
    *   **–ß—Ç–æ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç:**  MAE –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Å—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–º–µ—Ä "–∞–±—Å–æ–ª—é—Ç–Ω–æ–π" –æ—à–∏–±–∫–∏ –º–æ–¥–µ–ª–∏.  –ß–µ–º –º–µ–Ω—å—à–µ MAE, —Ç–µ–º –ª—É—á—à–µ –º–æ–¥–µ–ª—å.  MAE –º–µ–Ω–µ–µ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–∞ –∫ *–≤—ã–±—Ä–æ—Å–∞–º* –∏ *–±–æ–ª—å—à–∏–º –æ—à–∏–±–∫–∞–º*, —á–µ–º MSE, —Ç–∞–∫ –∫–∞–∫ –∞–±—Å–æ–ª—é—Ç–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ "—Å–≥–ª–∞–∂–∏–≤–∞–µ—Ç" –±–æ–ª—å—à–∏–µ –æ—à–∏–±–∫–∏, –Ω–µ —Ç–∞–∫ —Å–∏–ª—å–Ω–æ –∏—Ö —É—Å–∏–ª–∏–≤–∞—è, –∫–∞–∫ –∫–≤–∞–¥—Ä–∞—Ç.  MAE —Ö–æ—Ä–æ—à–æ –ø–æ–¥—Ö–æ–¥–∏—Ç, –∫–æ–≥–¥–∞ –≤–∞–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å *—Å—Ä–µ–¥–Ω—é—é* –æ—à–∏–±–∫—É –≤ *–∏—Å—Ö–æ–¥–Ω—ã—Ö –µ–¥–∏–Ω–∏—Ü–∞—Ö –∏–∑–º–µ—Ä–µ–Ω–∏—è*.
    *   **–ï–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è:**  –ï–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è MAE ‚Äì —Å–æ–≤–ø–∞–¥–∞—é—Ç —Å –µ–¥–∏–Ω–∏—Ü–∞–º–∏ –∏–∑–º–µ—Ä–µ–Ω–∏—è —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π.  –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ —Ü–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è ‚Äì —Ü–µ–Ω–∞ –≤ –¥–æ–ª–ª–∞—Ä–∞—Ö, —Ç–æ MAE –±—É–¥–µ—Ç –≤ "–¥–æ–ª–ª–∞—Ä–∞—Ö", —á—Ç–æ –±–æ–ª–µ–µ –∏–Ω—Ç—É–∏—Ç–∏–≤–Ω–æ –ø–æ–Ω—è—Ç–Ω–æ, —á–µ–º "–¥–æ–ª–ª–∞—Ä—ã –≤ –∫–≤–∞–¥—Ä–∞—Ç–µ" —É MSE.
    *   **–§–æ—Ä–º—É–ª–∞ MAE:**
    $$\frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$

    * –ì–¥–µ:
        - n ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π,
        - $y_i$ ‚Äî –∏—Å—Ç–∏–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ,
        - $\hat{y}_i$ ‚Äî –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ.


3.  **Mean Absolute Percentage Error (MAPE) ‚Äì –°—Ä–µ–¥–Ω—è—è –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–∞—è –æ—à–∏–±–∫–∞:** ÔºÖ
    *   **–ò–Ω—Ç—É–∏—Ç–∏–≤–Ω–æ:**  –°—Ä–µ–¥–Ω–∏–π –ø—Ä–æ—Ü–µ–Ω—Ç –æ—à–∏–±–∫–∏ –º–æ–¥–µ–ª–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ *—Ä–µ–∞–ª—å–Ω—ã—Ö* –∑–Ω–∞—á–µ–Ω–∏–π.
    *   **–ö–∞–∫ —Å—á–∏—Ç–∞–µ—Ç—Å—è:**  –î–ª—è –∫–∞–∂–¥–æ–π —Ç–æ—á–∫–∏ –¥–∞–Ω–Ω—ã—Ö –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –æ—à–∏–±–∫–∞ ‚Äì  –∞–±—Å–æ–ª—é—Ç–Ω–∞—è —Ä–∞–∑–Ω–æ—Å—Ç—å –º–µ–∂–¥—É –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º –∏ —Ä–µ–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º, –¥–µ–ª–µ–Ω–Ω–∞—è –Ω–∞ *—Ä–µ–∞–ª—å–Ω–æ–µ* –∑–Ω–∞—á–µ–Ω–∏–µ.  –ó–∞—Ç–µ–º —ç—Ç–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –æ—à–∏–±–∫–∞ –≤—ã—Ä–∞–∂–∞–µ—Ç—Å—è –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö (—É–º–Ω–æ–∂–∞–µ—Ç—Å—è –Ω–∞ 100%).  –ò —ç—Ç–∏ –ø—Ä–æ—Ü–µ–Ω—Ç–Ω—ã–µ –æ—à–∏–±–∫–∏ —É—Å—Ä–µ–¥–Ω—è—é—Ç—Å—è.
    *   **–ß—Ç–æ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç:**  MAPE –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Å—Ä–µ–¥–Ω–∏–π –ø—Ä–æ—Ü–µ–Ω—Ç –æ—à–∏–±–∫–∏ –º–æ–¥–µ–ª–∏.  –ß–µ–º –º–µ–Ω—å—à–µ MAPE, —Ç–µ–º –ª—É—á—à–µ –º–æ–¥–µ–ª—å.  MAPE –ª–µ–≥–∫–æ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å ‚Äì  –Ω–∞–ø—Ä–∏–º–µ—Ä, MAPE = 5% –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –≤ —Å—Ä–µ–¥–Ω–µ–º –º–æ–¥–µ–ª—å –æ—à–∏–±–∞–µ—Ç—Å—è –Ω–∞ 5% –æ—Ç —Ä–µ–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è.  MAPE —É–¥–æ–±–Ω–∞ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —à–∫–∞–ª–∞—Ö –∏–∑–º–µ—Ä–µ–Ω–∏–π, —Ç–∞–∫ –∫–∞–∫ —ç—Ç–æ –±–µ–∑—Ä–∞–∑–º–µ—Ä–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ (–ø—Ä–æ—Ü–µ–Ω—Ç).  –í–∞–∂–Ω–æ:  MAPE –Ω–µ–ª—å–∑—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å, –µ—Å–ª–∏ –≤ –¥–∞–Ω–Ω—ã—Ö –µ—Å—Ç—å *–Ω—É–ª–µ–≤—ã–µ* –∏–ª–∏ *–æ—á–µ–Ω—å –±–ª–∏–∑–∫–∏–µ –∫ –Ω—É–ª—é* —Ä–µ–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è, —Ç–∞–∫ –∫–∞–∫ –¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å –∏–ª–∏ –æ—á–µ–Ω—å –º–∞–ª–µ–Ω—å–∫–æ–µ —á–∏—Å–ª–æ –ø—Ä–∏–≤–µ–¥–µ—Ç –∫ –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–º –∏–ª–∏ –æ—á–µ–Ω—å –±–æ–ª—å—à–∏–º –∑–Ω–∞—á–µ–Ω–∏—è–º MAPE.
    *   **–ï–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è:**  –ü—Ä–æ—Ü–µ–Ω—Ç—ã (%) ‚Äì –±–µ–∑—Ä–∞–∑–º–µ—Ä–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞.


–í—Å–µ —ç—Ç–∏ –º–µ—Ç—Ä–∏–∫–∏ (MSE, MAE, MAPE) –ø–æ —Å—É—Ç–∏ –∏–∑–º–µ—Ä—è—é—Ç **"–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ" –∏–ª–∏ "–æ—à–∏–±–∫—É"** –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –º–æ–¥–µ–ª–∏ –æ—Ç —Ä–µ–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π.  –ß–µ–º **–º–µ–Ω—å—à–µ** —ç—Ç–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ, —Ç–µ–º **–ª—É—á—à–µ** –º–æ–¥–µ–ª—å.  –†–∞–∑–Ω–∏—Ü–∞ ‚Äì –≤ —Ç–æ–º, **–∫–∞–∫ –∏–º–µ–Ω–Ω–æ —ç—Ç–æ "–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ" –∏–∑–º–µ—Ä—è–µ—Ç—Å—è –∏ —É—Å—Ä–µ–¥–Ω—è–µ—Ç—Å—è**:

*   **MSE ‚Äì "–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ" (—Å–∏–ª—å–Ω–æ —à—Ç—Ä–∞—Ñ—É–µ—Ç –∑–∞ –±–æ–ª—å—à–∏–µ –æ—à–∏–±–∫–∏)**
*   **MAE ‚Äì "–∞–±—Å–æ–ª—é—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ" (–±–æ–ª–µ–µ —É—Å—Ç–æ–π—á–∏–≤–∞ –∫ –≤—ã–±—Ä–æ—Å–∞–º)**
*   **MAPE ‚Äì "–ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ" (–ª–µ–≥–∫–æ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å, –Ω–æ –æ—Å—Ç–æ—Ä–æ–∂–Ω–æ —Å –Ω—É–ª—è–º–∏)**

**–û—Å–Ω–æ–≤–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:**

#### **Accuracy (–¢–æ—á–Ω–æ—Å—Ç—å)**

**Accuracy** ‚Äì  —Å–∞–º–∞—è –ø—Ä–æ—Å—Ç–∞—è –∏, –≤–æ–∑–º–æ–∂–Ω–æ, —Å–∞–º–∞—è –∏–Ω—Ç—É–∏—Ç–∏–≤–Ω–æ –ø–æ–Ω—è—Ç–Ω–∞—è –º–µ—Ç—Ä–∏–∫–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏.  –û–Ω–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç **–¥–æ–ª—é *–≤–µ—Ä–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö* –æ–±—ä–µ–∫—Ç–æ–≤ –æ—Ç *–æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞* –æ–±—ä–µ–∫—Ç–æ–≤.**

**–§–æ—Ä–º—É–ª–∞ Accuracy:**

$ {Accuracy} = \frac{\text{–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–µ—Ä–Ω—ã—Ö}}{\text{–û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ}}$ 

–ò–ª–∏, –≤ —Ç–µ—Ä–º–∏–Ω–∞—Ö **–º–∞—Ç—Ä–∏—Ü—ã –æ—à–∏–±–æ–∫ (Confusion Matrix):**


$ {Accuracy} = \frac{\text{TP + TN}}{\text{TP + TN + FP + FN}}$ 


–ì–¥–µ:

*   **TP (True Positives) ‚Äì –ò—Å—Ç–∏–Ω–Ω–æ-–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ —Å—Ä–∞–±–∞—Ç—ã–≤–∞–Ω–∏—è:**  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–¥–µ–ª—å **–≤–µ—Ä–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–ª–∞ –∫–∞–∫ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å*** (–Ω–∞–ø—Ä–∏–º–µ—Ä, –≤–µ—Ä–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏–ª–∞, —á—Ç–æ –ø–∏—Å—å–º–æ ‚Äì —Å–ø–∞–º).
*   **TN (True Negatives) ‚Äì –ò—Å—Ç–∏–Ω–Ω–æ-–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ —Å—Ä–∞–±–∞—Ç—ã–≤–∞–Ω–∏—è:**  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–¥–µ–ª—å **–≤–µ—Ä–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–ª–∞ –∫–∞–∫ *–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å*** (–Ω–∞–ø—Ä–∏–º–µ—Ä, –≤–µ—Ä–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏–ª–∞, —á—Ç–æ –ø–∏—Å—å–º–æ ‚Äì –Ω–µ —Å–ø–∞–º).
*   **FP (False Positives) ‚Äì –õ–æ–∂–Ω–æ-–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ —Å—Ä–∞–±–∞—Ç—ã–≤–∞–Ω–∏—è (–æ—à–∏–±–∫–∏ I —Ä–æ–¥–∞):**  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–¥–µ–ª—å **–æ—à–∏–±–æ—á–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–ª–∞ –∫–∞–∫ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å***, —Ö–æ—Ç—è –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ –æ–Ω–∏ –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫ *–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–º—É –∫–ª–∞—Å—Å—É* (–Ω–∞–ø—Ä–∏–º–µ—Ä, –æ—à–∏–±–æ—á–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏–ª–∞, —á—Ç–æ –ø–∏—Å—å–º–æ ‚Äì —Å–ø–∞–º, —Ö–æ—Ç—è –æ–Ω–æ –Ω–µ —Å–ø–∞–º ‚Äì "–ª–æ–∂–Ω–∞—è —Ç—Ä–µ–≤–æ–≥–∞").
*   **FN (False Negatives) ‚Äì –õ–æ–∂–Ω–æ-–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ —Å—Ä–∞–±–∞—Ç—ã–≤–∞–Ω–∏—è (–æ—à–∏–±–∫–∏ II —Ä–æ–¥–∞):**  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–¥–µ–ª—å **–æ—à–∏–±–æ—á–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–ª–∞ –∫–∞–∫ *–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å***, —Ö–æ—Ç—è –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ –æ–Ω–∏ –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–º—É –∫–ª–∞—Å—Å—É* (–Ω–∞–ø—Ä–∏–º–µ—Ä, –æ—à–∏–±–æ—á–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏–ª–∞, —á—Ç–æ –ø–∏—Å—å–º–æ ‚Äì –Ω–µ —Å–ø–∞–º, —Ö–æ—Ç—è –æ–Ω–æ —Å–ø–∞–º ‚Äì "–ø—Ä–æ–ø—É—Å–∫ —Ü–µ–ª–∏").

Accuracy –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ —á–∞—Å—Ç–æ –º–æ–¥–µ–ª—å *–≤ —Ü–µ–ª–æ–º* –ø—Ä–∞–≤–∏–ª—å–Ω–æ —É–≥–∞–¥—ã–≤–∞–µ—Ç –∫–ª–∞—Å—Å—ã –æ–±—ä–µ–∫—Ç–æ–≤.  –ß–µ–º –≤—ã—à–µ Accuracy, —Ç–µ–º –ª—É—á—à–µ –º–æ–¥–µ–ª—å.  **–ò–¥–µ–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ Accuracy ‚Äì 1.0 (–∏–ª–∏ 100%)**, —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –º–æ–¥–µ–ª—å –∏–¥–µ–∞–ª—å–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä—É–µ—Ç –≤—Å–µ –æ–±—ä–µ–∫—Ç—ã (—á—Ç–æ –Ω–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ –ø–æ—á—Ç–∏ –Ω–∏–∫–æ–≥–¥–∞ –Ω–µ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è).

#### **Presicion**

**Precision ‚Äì  –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, *–∫–∞–∫—É—é –¥–æ–ª—é –∏–∑ –æ–±—ä–µ–∫—Ç–æ–≤, *–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö* –º–æ–¥–µ–ª—å—é –∫–∞–∫ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å***, –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ —è–≤–ª—è—é—Ç—Å—è *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º –∫–ª–∞—Å—Å–æ–º*. Precision –æ—Ç–≤–µ—á–∞–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å: "–ù–∞—Å–∫–æ–ª—å–∫–æ –º—ã –º–æ–∂–µ–º –¥–æ–≤–µ—Ä—è—Ç—å positive predictions –º–æ–¥–µ–ª–∏?"** –∏–ª–∏ "–ö–∞–∫ –º–Ω–æ–≥–æ –∏–∑ *–ø–æ–º–µ—á–µ–Ω–Ω—ã—Ö* –º–æ–¥–µ–ª—å—é –∫–∞–∫ '—Å–ø–∞–º' –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ —è–≤–ª—è—é—Ç—Å—è —Å–ø–∞–º–æ–º?'"

**–§–æ—Ä–º—É–ª–∞ Precision:**  

${Precision} = \frac{\text{TP}}{\text{TP + FP}}$

*   **TP (True Positives) ‚Äì  –ß–∏—Å–ª–∏—Ç–µ–ª—å:** –í–µ—Ä–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∫–∞–∫ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å.  –≠—Ç–æ "—É—Å–ø–µ—Ö" –º–æ–¥–µ–ª–∏ –≤ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞.
*   **TP + FP (True Positives + False Positives) ‚Äì  –ó–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å:**  –í—Å–µ –æ–±—ä–µ–∫—Ç—ã, *–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ* –º–æ–¥–µ–ª—å—é –∫–∞–∫ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å.  –≠—Ç–æ –≤—Å–µ "–ø–æ–∑–∏—Ç–∏–≤–Ω—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è" –º–æ–¥–µ–ª–∏, –∫–∞–∫ –≤–µ—Ä–Ω—ã–µ (TP), —Ç–∞–∫ –∏ –æ—à–∏–±–æ—á–Ω—ã–µ (FP).

**Precision ‚Äì  —ç—Ç–æ "—Ç–æ—á–Ω–æ—Å—Ç—å" –∏–ª–∏ "–∞–∫–∫—É—Ä–∞—Ç–Ω–æ—Å—Ç—å" *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π* –º–æ–¥–µ–ª–∏.**  –í—ã—Å–æ–∫–∏–π Precision –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –º–æ–¥–µ–ª—å –¥–µ–ª–∞–µ—Ç *–º–∞–ª–æ –ª–æ–∂–Ω–æ-–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –æ—à–∏–±–æ–∫ (FP)*.  –¢–æ –µ—Å—Ç—å, –∫–æ–≥–¥–∞ –º–æ–¥–µ–ª—å –≥–æ–≤–æ—Ä–∏—Ç "—ç—Ç–æ ‚Äì –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å", –æ–Ω–∞ —á–∞—Å—Ç–æ –±—ã–≤–∞–µ—Ç –ø—Ä–∞–≤–∞.

**–ü—Ä–∏–º–µ—Ä:**
|                | –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–π: –°–ø–∞–º | –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–π: –ù–µ-—Å–ø–∞–º |
| :------------- | :-----------------: | :--------------------: |
| –ò—Å—Ç–∏–Ω–Ω—ã–π: –°–ø–∞–º |        **TP = 20**       |        **FN = 5**        |
| –ò—Å—Ç–∏–Ω–Ω—ã–π: –ù–µ-—Å–ø–∞–º |       **FP = 5**        |       **TN = 70**       |

**–†–∞—Å—á–µ—Ç Precision:**

${Precision} = \frac{\text{TP}}{\text{TP + FP}} = \frac{20}{20 + 5} = \frac{20}{25} = 0.80$

–í —ç—Ç–æ–º –ø—Ä–∏–º–µ—Ä–µ Precision = 80%.  –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –∏–∑ –≤—Å–µ—Ö –ø–∏—Å–µ–º, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–¥–µ–ª—å *–ø–æ–º–µ—Ç–∏–ª–∞ –∫–∞–∫ —Å–ø–∞–º*, 80% –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –æ–∫–∞–∑–∞–ª–∏—Å—å —Å–ø–∞–º–æ–º.  –û—Å—Ç–∞–ª—å–Ω—ã–µ 20% (FP = 5 –∏–∑ 25) ‚Äì —ç—Ç–æ "–ª–æ–∂–Ω—ã–µ —Å—Ä–∞–±–∞—Ç—ã–≤–∞–Ω–∏—è" (–Ω–µ-—Å–ø–∞–º –ø–∏—Å—å–º–∞, –æ—à–∏–±–æ—á–Ω–æ –ø–æ–º–µ—á–µ–Ω–Ω—ã–µ –∫–∞–∫ —Å–ø–∞–º).

#### **Recall (–ü–æ–ª–Ω–æ—Ç–∞, –ß—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å, True Positive Rate)**

**Recall ‚Äì  –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, *–∫–∞–∫—É—é –¥–æ–ª—é –∏–∑ –æ–±—ä–µ–∫—Ç–æ–≤, *–¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ* –æ—Ç–Ω–æ—Å—è—â–∏—Ö—Å—è –∫ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–º—É –∫–ª–∞—Å—Å—É***, –º–æ–¥–µ–ª—å *–≤–µ—Ä–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–ª–∞* –∫–∞–∫ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å*.**  Recall –æ—Ç–≤–µ—á–∞–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å: "–ù–∞—Å–∫–æ–ª—å–∫–æ –ø–æ–ª–Ω–æ –º–æ–¥–µ–ª—å *–æ–±–Ω–∞—Ä—É–∂–∏–≤–∞–µ—Ç* positive cases?" –∏–ª–∏ "–ö–∞–∫ –º–Ω–æ–≥–æ *—Ä–µ–∞–ª—å–Ω–æ–≥–æ* —Å–ø–∞–º–∞ –º–æ–¥–µ–ª—å *–ø–æ–π–º–∞–ª–∞* –∏–∑ *–≤—Å–µ–≥–æ* —Å–ø–∞–º–∞?"

**–§–æ—Ä–º—É–ª–∞ Recall:**

${Recall} = \frac{\text{TP}}{\text{TP + FN}}$

*   **TP (True Positives) ‚Äì  –ß–∏—Å–ª–∏—Ç–µ–ª—å:**  –í–µ—Ä–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∫–∞–∫ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å.  –≠—Ç–æ "—É—Å–ø–µ—Ö" –º–æ–¥–µ–ª–∏ –≤ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞.
*   **TP + FN (True Positives + False Negatives) ‚Äì  –ó–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å:**  –í—Å–µ –æ–±—ä–µ–∫—Ç—ã, –∫–æ—Ç–æ—Ä—ã–µ *–Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ* –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–º—É –∫–ª–∞—Å—Å—É.  –≠—Ç–æ –≤—Å–µ "—Ä–µ–∞–ª—å–Ω—ã–µ positive cases", –∫–∞–∫ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–Ω—ã–µ –º–æ–¥–µ–ª—å—é (TP), —Ç–∞–∫ –∏ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –º–æ–¥–µ–ª—å—é (FN).

**Recall ‚Äì  —ç—Ç–æ "–ø–æ–ª–Ω–æ—Ç–∞", "—á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å" –∏–ª–∏ "–¥–æ–ª—è *–æ–±–Ω–∞—Ä—É–∂–µ–Ω–Ω—ã—Ö* –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤ –æ—Ç *–≤—Å–µ—Ö* –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤.**  –í—ã—Å–æ–∫–∏–π Recall –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –º–æ–¥–µ–ª—å –¥–µ–ª–∞–µ—Ç *–º–∞–ª–æ –ª–æ–∂–Ω–æ-–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö –æ—à–∏–±–æ–∫ (FN)*.  –¢–æ –µ—Å—Ç—å, –º–æ–¥–µ–ª—å *–Ω–µ –ø—Ä–æ–ø—É—Å–∫–∞–µ—Ç* –º–Ω–æ–≥–æ —Ä–µ–∞–ª—å–Ω—ã—Ö –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤.

**–ü—Ä–∏–º–µ—Ä:**
|                | –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–π: –°–ø–∞–º | –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–π: –ù–µ-—Å–ø–∞–º |
| :------------- | :-----------------: | :--------------------: |
| –ò—Å—Ç–∏–Ω–Ω—ã–π: –°–ø–∞–º |        **TP = 20**       |        **FN = 5**        |
| –ò—Å—Ç–∏–Ω–Ω—ã–π: –ù–µ-—Å–ø–∞–º |       **FP = 5**        |       **TN = 70**       |

**–†–∞—Å—á–µ—Ç Recall:**

${Recall} = \frac{\text{TP}}{\text{TP + FN}} = \frac{20}{20 + 5} = \frac{20}{25} = 0.8$

**–í —ç—Ç–æ–º –ø—Ä–∏–º–µ—Ä–µ Recall = 80%.**  –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –∏–∑ *–≤—Å–µ—Ö* —Å–ø–∞–º-–ø–∏—Å–µ–º, –∫–æ—Ç–æ—Ä—ã–µ *–±—ã–ª–∏ –≤ –Ω–∞–±–æ—Ä–µ –¥–∞–Ω–Ω—ã—Ö*, –º–æ–¥–µ–ª—å *–æ–±–Ω–∞—Ä—É–∂–∏–ª–∞* 80% –∏–∑ –Ω–∏—Ö.  –û—Å—Ç–∞–ª—å–Ω—ã–µ 20% (FN = 5 –∏–∑ 25) ‚Äì —ç—Ç–æ —Å–ø–∞–º-–ø–∏—Å—å–º–∞, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–¥–µ–ª—å –ø—Ä–æ–ø—É—Å—Ç–∏–ª–∞ –∏ –æ—à–∏–±–æ—á–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–ª–∞ –∫–∞–∫ –Ω–µ-—Å–ø–∞–º.

**–í–∞–∂–Ω–æ:**

*   **Precision –∏ Recall ‚Äì  –º–µ—Ç—Ä–∏–∫–∏, –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –Ω–∞ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å*.**  –û–Ω–∏ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç –∫–∞—á–µ—Å—Ç–≤–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∏–º–µ–Ω–Ω–æ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞.  –î–ª—è –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞ –µ—Å—Ç—å –∞–Ω–∞–ª–æ–≥–∏—á–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏, –Ω–æ Precision –∏ Recall –æ–±—ã—á–Ω–æ –∏—Å–ø–æ–ª—å–∑—É—é—Ç –∏–º–µ–Ω–Ω–æ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞, –æ—Å–æ–±–µ–Ω–Ω–æ –≤ –∑–∞–¥–∞—á–∞—Ö, –≥–¥–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å —è–≤–ª—è–µ—Ç—Å—è *—Ü–µ–ª–µ–≤—ã–º* –∏ *–±–æ–ª–µ–µ –≤–∞–∂–Ω—ã–º* –¥–ª—è –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è.
*   **Precision –∏ Recall *—Ç–æ—Ä–≥—É—é—Ç* –¥—Ä—É–≥ —Å –¥—Ä—É–≥–æ–º (Precision-Recall Tradeoff).**  –û–±—ã—á–Ω–æ, –ø—Ä–∏ –ø–æ–ø—ã—Ç–∫–µ —É–≤–µ–ª–∏—á–∏—Ç—å Precision, Recall *—É–º–µ–Ω—å—à–∞–µ—Ç—Å—è*, –∏ –Ω–∞–æ–±–æ—Ä–æ—Ç.  –ü–æ—á–µ–º—É —Ç–∞–∫ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç?  –ü—Ä–µ–¥—Å—Ç–∞–≤—å —Å–µ–±–µ "–æ—á–µ–Ω—å —Å—Ç—Ä–æ–≥–∏–π" —Å–ø–∞–º-—Ñ–∏–ª—å—Ç—Ä, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–º–µ—á–∞–µ—Ç –∫–∞–∫ —Å–ø–∞–º —Ç–æ–ª—å–∫–æ *—Å–∞–º—ã–µ –æ—á–µ–≤–∏–¥–Ω—ã–µ* —Å–ø–∞–º-–ø–∏—Å—å–º–∞, –≤ –∫–æ—Ç–æ—Ä—ã—Ö —É–≤–µ—Ä–µ–Ω –Ω–∞ 100%.  –¢–∞–∫–æ–π —Ñ–∏–ª—å—Ç—Ä –±—É–¥–µ—Ç –∏–º–µ—Ç—å –≤—ã—Å–æ–∫–∏–π Precision (–ø–æ—á—Ç–∏ –≤—Å–µ –ø–æ–º–µ—á–µ–Ω–Ω—ã–µ –∫–∞–∫ —Å–ø–∞–º –ø–∏—Å—å–º–∞ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –±—É–¥—É—Ç —Å–ø–∞–º–æ–º), –Ω–æ –ø—Ä–∏ —ç—Ç–æ–º –º–æ–∂–µ—Ç –∏–º–µ—Ç—å –Ω–∏–∑–∫–∏–π Recall (–ø—Ä–æ–ø—É—Å–∫–∞—Ç—å –º–Ω–æ–≥–æ "–º–µ–Ω–µ–µ –æ—á–µ–≤–∏–¥–Ω–æ–≥–æ" —Å–ø–∞–º–∞).  –ò –Ω–∞–æ–±–æ—Ä–æ—Ç, "–æ—á–µ–Ω—å "–º—è–≥–∫–∏–π" —Å–ø–∞–º-—Ñ–∏–ª—å—Ç—Ä, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–º–µ—á–∞–µ—Ç –∫–∞–∫ —Å–ø–∞–º *–≤—Å–µ –ø–æ–¥–æ–∑—Ä–∏—Ç–µ–ª—å–Ω—ã–µ* –ø–∏—Å—å–º–∞, –¥–∞–∂–µ –µ—Å–ª–∏ –Ω–µ —É–≤–µ—Ä–µ–Ω –¥–æ –∫–æ–Ω—Ü–∞.  –¢–∞–∫–æ–π —Ñ–∏–ª—å—Ç—Ä –±—É–¥–µ—Ç –∏–º–µ—Ç—å **–≤—ã—Å–æ–∫–∏–π Recall** (–ø–æ—á—Ç–∏ –≤–µ—Å—å —Å–ø–∞–º –±—É–¥–µ—Ç –ø–æ–π–º–∞–Ω), –Ω–æ –ø—Ä–∏ —ç—Ç–æ–º –Ω–∏–∑–∫–∏–π Precision (–º–Ω–æ–≥–æ –Ω–µ-—Å–ø–∞–º–∞ –±—É–¥–µ—Ç –æ—à–∏–±–æ—á–Ω–æ –ø–æ–º–µ—á–µ–Ω–æ –∫–∞–∫ —Å–ø–∞–º - "–ª–æ–∂–Ω—ã–µ —Å—Ä–∞–±–∞—Ç—ã–≤–∞–Ω–∏—è").

#### **F1-score**

**F1-score (F1-–º–µ—Ä–∞)** ‚Äì  —ç—Ç–æ **–≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –º–µ–∂–¥—É Precision –∏ Recall.**   

**–ü–æ—á–µ–º—É "–≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ"?**

–ö–æ–≥–¥–∞ –º—ã —Ö–æ—Ç–∏–º —É—Å—Ä–µ–¥–Ω–∏—Ç—å –¥–≤–∞ —á–∏—Å–ª–∞, –º—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º **–∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ**: (a + b) / 2.  –ù–æ –¥–ª—è Precision –∏ Recall –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –Ω–µ –æ—á–µ–Ω—å –ø–æ–¥—Ö–æ–¥–∏—Ç, –ø–æ—Ç–æ–º—É —á—Ç–æ –æ–Ω–æ —Å–ª–∏—à–∫–æ–º "–º—è–≥–∫–æ–µ" –∏ –º–æ–∂–µ—Ç "–ø—Ä–æ—â–∞—Ç—å" —Å–∏—Ç—É–∞—Ü–∏–∏, –∫–æ–≥–¥–∞ –æ–¥–Ω–∞ –∏–∑ –º–µ—Ç—Ä–∏–∫ –æ—á–µ–Ω—å –Ω–∏–∑–∫–∞—è, –∞ –¥—Ä—É–≥–∞—è –≤—ã—Å–æ–∫–∞—è.

**–ü—Ä–∏–º–µ—Ä:**

–ö –ø—Ä–∏–º–µ—Ä—É, —É –Ω–∞—Å –µ—Å—Ç—å –¥–≤–µ –º–æ–¥–µ–ª–∏:

*   **–ú–æ–¥–µ–ª—å 1:** Precision = 90%, Recall = 50%
*   **–ú–æ–¥–µ–ª—å 2:** Precision = 60%, Recall = 80%

**–ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ:**

*   –ú–æ–¥–µ–ª—å 1: (90% + 50%) / 2 = 70%
*   –ú–æ–¥–µ–ª—å 2: (60% + 80%) / 2 = 70%

–ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –¥–ª—è –æ–±–µ–∏—Ö –º–æ–¥–µ–ª–µ–π ‚Äì **–æ–¥–∏–Ω–∞–∫–æ–≤–æ–µ (70%)!**  –ù–æ **–≤–∏–∑—É–∞–ª—å–Ω–æ, –º–æ–¥–µ–ª–∏ *–æ—á–µ–Ω—å —Ä–∞–∑–Ω—ã–µ***!  –ú–æ–¥–µ–ª—å 1 ‚Äì  –æ—á–µ–Ω—å —Ç–æ—á–Ω–∞—è (–≤—ã—Å–æ–∫–∏–π Precision), –Ω–æ –ø—Ä–æ–ø—É—Å–∫–∞–µ—Ç –º–Ω–æ–≥–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤ (–Ω–∏–∑–∫–∏–π Recall).  –ú–æ–¥–µ–ª—å 2 ‚Äì  –º–µ–Ω–µ–µ —Ç–æ—á–Ω–∞—è (–Ω–∏–∂–µ Precision), –Ω–æ –ª–æ–≤–∏—Ç –ø–æ—á—Ç–∏ –≤—Å–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ —Å–ª—É—á–∞–∏ (–≤—ã—Å–æ–∫–∏–π Recall).  –ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –Ω–µ "—á—É–≤—Å—Ç–≤—É–µ—Ç" —ç—Ç—É —Ä–∞–∑–Ω–∏—Ü—É –≤ –±–∞–ª–∞–Ω—Å–µ –º–µ–∂–¥—É Precision –∏ Recall.

**–ì–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ** ‚Äì  –±–æ–ª–µ–µ "—Å—Ç—Ä–æ–≥–æ–µ" –∏ "—á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ–µ" –∫ *–¥–∏—Å–±–∞–ª–∞–Ω—Å—É* –º–µ–∂–¥—É Precision –∏ Recall.  –û–Ω–æ "—à—Ç—Ä–∞—Ñ—É–µ—Ç" –º–æ–¥–µ–ª—å, –µ—Å–ª–∏ –æ–¥–Ω–∞ –∏–∑ –º–µ—Ç—Ä–∏–∫ (Precision –∏–ª–∏ Recall) —Å–∏–ª—å–Ω–æ –ø—Ä–æ—Å–µ–¥–∞–µ—Ç, –¥–∞–∂–µ –µ—Å–ª–∏ –¥—Ä—É–≥–∞—è –ø—Ä–∏ —ç—Ç–æ–º –æ—Å—Ç–∞–µ—Ç—Å—è –≤—ã—Å–æ–∫–æ–π.

**–§–æ—Ä–º—É–ª–∞ F1-score (–≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ):**

${F1\text{-}score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$

–ò–ª–∏, —á—Ç–æ —Ç–æ –∂–µ —Å–∞–º–æ–µ, –Ω–æ –±–æ–ª–µ–µ —Ä–∞–∑–≤–µ—Ä–Ω—É—Ç–æ, —á–µ—Ä–µ–∑ TP, FP, FN:

${F1\text{-}score} = \frac{2 \times \text{TP}}{2 \times \text{TP} + \text{FP} + \text{FN}}$

*   **Precision –∏ Recall ‚Äì  —É—á–∞—Å—Ç–≤—É—é—Ç –≤ —Ñ–æ—Ä–º—É–ª–µ –∫–∞–∫ *–º–Ω–æ–∂–∏—Ç–µ–ª–∏* –≤ —á–∏—Å–ª–∏—Ç–µ–ª–µ –∏ –∫–∞–∫ *—Å—É–º–º–∞* –≤ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª–µ.**  –≠—Ç–æ –∏ –¥–µ–ª–∞–µ—Ç F1-score –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–∏–º —Å—Ä–µ–¥–Ω–∏–º.
*   **F1-score –≤—Å–µ–≥–¥–∞ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –º–µ–∂–¥—É 0 –∏ 1 (–∏–ª–∏ 0% –∏ 100%)**.  –ß–µ–º –±–ª–∏–∂–µ F1-score –∫ 1 (–∏–ª–∏ 100%), —Ç–µ–º –ª—É—á—à–µ –º–æ–¥–µ–ª—å.  –ò–¥–µ–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ F1-score ‚Äì 1, —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –º–æ–¥–µ–ª—å –∏–º–µ–µ—Ç –∏ Precision = 1, –∏ Recall = 1 (–∏–¥–µ–∞–ª—å–Ω–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å –∏ –ø–æ–ª–Ω–æ—Ç–∞).  –ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ F1-score ‚Äì 0.

**F1-score ‚Äì  —ç—Ç–æ –º–µ—Ç—Ä–∏–∫–∞, –∫–æ—Ç–æ—Ä–∞—è —Å—Ç—Ä–µ–º–∏—Ç—Å—è –Ω–∞–π—Ç–∏ *–±–∞–ª–∞–Ω—Å* –º–µ–∂–¥—É Precision –∏ Recall.**  –í—ã—Å–æ–∫–∏–π F1-score –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ —Ç–æ–≥–¥–∞, –∫–æ–≥–¥–∞ –∏ Precision, –∏ Recall *–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –≤—ã—Å–æ–∫–∏* –∏ *–±–ª–∏–∑–∫–∏ –¥—Ä—É–≥ –∫ –¥—Ä—É–≥—É*.  –ï—Å–ª–∏ –æ–¥–Ω–∞ –∏–∑ –º–µ—Ç—Ä–∏–∫ (Precision –∏–ª–∏ Recall) —Å–∏–ª—å–Ω–æ –ø—Ä–æ—Å–µ–¥–∞–µ—Ç, F1-score —Ç–æ–∂–µ –±—É–¥–µ—Ç *–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –Ω–∏–∂–µ*, –¥–∞–∂–µ –µ—Å–ª–∏ –¥—Ä—É–≥–∞—è –º–µ—Ç—Ä–∏–∫–∞ –ø—Ä–∏ —ç—Ç–æ–º –æ—Å—Ç–∞–µ—Ç—Å—è –≤—ã—Å–æ–∫–æ–π.

F1-score ‚Äì  –æ—Å–æ–±–µ–Ω–Ω–æ –ø–æ–ª–µ–∑–Ω–∞ –≤ –∑–∞–¥–∞—á–∞—Ö —Å *–Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –∫–ª–∞—Å—Å–∞–º–∏*, –∏ –∫–æ–≥–¥–∞ *–≤–∞–∂–Ω—ã –∏ Precision, –∏ Recall*, –∏ –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ *–æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–π –±–∞–ª–∞–Ω—Å* –º–µ–∂–¥—É –Ω–∏–º–∏.  –ù–∞–ø—Ä–∏–º–µ—Ä, –≤ –∑–∞–¥–∞—á–∞—Ö –º–µ–¥–∏—Ü–∏–Ω—Å–∫–æ–π –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∏, –≤–∞–∂–Ω–æ –∏ –Ω–µ –ø—Ä–æ–ø—É—Å—Ç–∏—Ç—å –±–æ–ª–µ–∑–Ω—å (–≤—ã—Å–æ–∫–∏–π Recall), –∏ –Ω–µ —Å—Ç–∞–≤–∏—Ç—å –ª–æ–∂–Ω—ã–π –¥–∏–∞–≥–Ω–æ–∑ –∑–¥–æ—Ä–æ–≤—ã–º –ª—é–¥—è–º (–≤—ã—Å–æ–∫–∏–π Precision).  F1-score –ø–æ–º–æ–≥–∞–µ—Ç –æ—Ü–µ–Ω–∏—Ç—å, –Ω–∞—Å–∫–æ–ª—å–∫–æ –º–æ–¥–µ–ª—å —Ö–æ—Ä–æ—à–æ —Å–ø—Ä–∞–≤–ª—è–µ—Ç—Å—è —Å –æ–±–µ–∏–º–∏ —ç—Ç–∏–º–∏ –∑–∞–¥–∞—á–∞–º–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ.

**–ü—Ä–∏–º–µ—Ä:**

–í–µ—Ä–Ω–µ–º—Å—è –∫ –Ω–∞—à–∏–º –¥–≤—É–º –º–æ–¥–µ–ª—è–º:

*   **–ú–æ–¥–µ–ª—å 1:** Precision = 90%, Recall = 50%
*   **–ú–æ–¥–µ–ª—å 2:** Precision = 60%, Recall = 80%

**–†–∞—Å—á–µ—Ç F1-score:**

**–ú–æ–¥–µ–ª—å 1:**  
${F1\text{-}score} = 2 \times \frac{0.9 \times 0.5}{0.9 + 0.5} = \frac{0.9}{1.4} \approx 0.64 \quad$

**–ú–æ–¥–µ–ª—å 2:**  
${F1\text{-}score} = 2 \times \frac{0.6 \times 0.8}{0.6 + 0.8} = \frac{0.96}{1.4} \approx 0.69$

–¢–µ–ø–µ—Ä—å F1-score –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, —á—Ç–æ –ú–æ–¥–µ–ª—å 2 *–ª—É—á—à–µ*, —á–µ–º –ú–æ–¥–µ–ª—å 1 (F1-score –≤—ã—à–µ –¥–ª—è –ú–æ–¥–µ–ª–∏ 2).  F1-score "—É–≤–∏–¥–µ–ª" —Ä–∞–∑–Ω–∏—Ü—É –≤ –±–∞–ª–∞–Ω—Å–µ –º–µ–∂–¥—É Precision –∏ Recall, –∫–æ—Ç–æ—Ä—É—é "–ø—Ä–æ–ø—É—Å—Ç–∏–ª–æ" –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ.  –ú–æ–¥–µ–ª—å 2, —Ö–æ—Ç—å –∏ –∏–º–µ–µ—Ç –Ω–µ–º–Ω–æ–≥–æ –Ω–∏–∂–µ Precision, —á–µ–º –ú–æ–¥–µ–ª—å 1, –Ω–æ –∑–∞—Ç–æ –∏–º–µ–µ—Ç –≥–æ—Ä–∞–∑–¥–æ –≤—ã—à–µ Recall, –∏ F1-score –æ—Ü–µ–Ω–∏–ª —ç—Ç–æ—Ç –±–∞–ª–∞–Ω—Å –∫–∞–∫ –±–æ–ª–µ–µ –ø—Ä–µ–¥–ø–æ—á—Ç–∏—Ç–µ–ª—å–Ω—ã–π.

**–í–∞–∂–Ω–æ:**

*   **F1-score ‚Äì  —ç—Ç–æ –º–µ—Ç—Ä–∏–∫–∞ –¥–ª—è *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞* (–∫–∞–∫ –∏ Precision –∏ Recall).**  –ï—Å–ª–∏ —É —Ç–µ–±—è –±–∏–Ω–∞—Ä–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è, –Ω—É–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å, –∫–∞–∫–æ–π –∫–ª–∞—Å—Å —Å—á–∏—Ç–∞—Ç—å "–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º", –∏ F1-score –±—É–¥–µ—Ç –æ—Ü–µ–Ω–∏–≤–∞—Ç—å –∫–∞—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∏–º–µ–Ω–Ω–æ –¥–ª—è —ç—Ç–æ–≥–æ –∫–ª–∞—Å—Å–∞.
*   **F1-score ‚Äì  —ç—Ç–æ *–æ–¥–Ω–∞ –º–µ—Ç—Ä–∏–∫–∞*, –∫–æ—Ç–æ—Ä–∞—è –æ–±—ä–µ–¥–∏–Ω—è–µ—Ç Precision –∏ Recall.**  –≠—Ç–æ —É–¥–æ–±–Ω–æ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π –ø–æ *–æ–±—â–µ–º—É –∫–∞—á–µ—Å—Ç–≤—É* –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏, –∫–æ–≥–¥–∞ –≤–∞–∂–µ–Ω –±–∞–ª–∞–Ω—Å –º–µ–∂–¥—É Precision –∏ Recall.  –ù–æ F1-score –Ω–µ –¥–∞–µ—Ç *–ø–æ–ª–Ω–æ–π –∫–∞—Ä—Ç–∏–Ω—ã* –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏.  –í –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö –∑–∞–¥–∞—á–∞—Ö –º–æ–∂–µ—Ç –±—ã—Ç—å –≤–∞–∂–Ω–æ —Å–º–æ—Ç—Ä–µ—Ç—å –Ω–∞ Precision –∏ Recall*–æ—Ç–¥–µ–ª—å–Ω–æ, —á—Ç–æ–±—ã –ª—É—á—à–µ –ø–æ–Ω—è—Ç—å, –∫–∞–∫–∏–µ —Ç–∏–ø—ã –æ—à–∏–±–æ–∫ –ø—Ä–µ–æ–±–ª–∞–¥–∞—é—Ç –∏ –∫–∞–∫ –∏—Ö –º–æ–∂–Ω–æ –∏—Å–ø—Ä–∞–≤–∏—Ç—å.

---

ROC AUC (Receiver Operating Characteristic Area Under the Curve) –∏ PR AUC (Precision-Recall Area Under the Curve) ‚Äì  —ç—Ç–æ –º–µ—Ç—Ä–∏–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–∑–≤–æ–ª—è—é—Ç –æ—Ü–µ–Ω–∏—Ç—å –∫–∞—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ *–≤ —Ü–µ–ª–æ–º*, *–Ω–µ –ø—Ä–∏–≤—è–∑—ã–≤–∞—è—Å—å –∫ –∫–∞–∫–æ–º—É-—Ç–æ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–º—É –ø–æ—Ä–æ–≥—É –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏*.

–í—Å–ø–æ–º–Ω–∏–º, –∫–∞–∫ —Ä–∞–±–æ—Ç–∞—é—Ç –º–æ–¥–µ–ª–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è).  –û–±—ã—á–Ω–æ –æ–Ω–∏ –Ω–µ –≤—ã–¥–∞—é—Ç —Å—Ä–∞–∑—É "–∫–∞—Ç–µ–≥–æ—Ä–∏—é" ("—Å–ø–∞–º" –∏–ª–∏ "–Ω–µ —Å–ø–∞–º"), –∞ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—é—Ç *–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å* –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –∫ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–º—É –∫–ª–∞—Å—Å—É* (–Ω–∞–ø—Ä–∏–º–µ—Ä, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ø–∏—Å—å–º–æ ‚Äì —Å–ø–∞–º).  –ê —É–∂–µ –ø–æ—Ç–æ–º, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å *–∫–æ–Ω–∫—Ä–µ—Ç–Ω—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏—é*, –º—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º –ø–æ—Ä–æ–≥ (threshold).  –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å > 0.5, —Ç–æ –æ—Ç–Ω–æ—Å–∏–º –∫ "—Å–ø–∞–º—É", –µ—Å–ª–∏ ‚â§ 0.5 ‚Äì –∫ "–Ω–µ-—Å–ø–∞–º—É".

–ü—Ä–æ–±–ª–µ–º–∞ –≤ —Ç–æ–º, —á—Ç–æ *–≤—ã–±–æ—Ä –ø–æ—Ä–æ–≥–∞* ‚Äì  —ç—Ç–æ *–ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ*, –∏ *–∫–∞—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –º–æ–∂–µ—Ç *—Å–∏–ª—å–Ω–æ –∑–∞–≤–∏—Å–µ—Ç—å –æ—Ç –≤—ã–±–æ—Ä–∞ –ø–æ—Ä–æ–≥–∞*.  –ï—Å–ª–∏ –º—ã –∏–∑–º–µ–Ω–∏–º –ø–æ—Ä–æ–≥, Precision –∏ Recall –∏–∑–º–µ–Ω—è—Ç—Å—è (–æ–±—ã—á–Ω–æ –≤ –ø—Ä–æ—Ç–∏–≤–æ–ø–æ–ª–æ–∂–Ω—ã—Ö –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è—Ö ‚Äì  –∫–∞–∫ "–∫–∞—á–µ–ª–∏ Precision-Recall Tradeoff").  –ö–∞–∫–æ–π –ø–æ—Ä–æ–≥ –≤—ã–±—Ä–∞—Ç—å ‚Äì  –Ω–µ –≤—Å–µ–≥–¥–∞ –æ—á–µ–≤–∏–¥–Ω–æ.

ROC AUC –∏ PR AUC ‚Äì  –ø–æ–º–æ–≥–∞—é—Ç –Ω–∞–º *–æ—Ü–µ–Ω–∏—Ç—å –º–æ–¥–µ–ª—å *–Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –≤—ã–±–æ—Ä–∞ –ø–æ—Ä–æ–≥–∞*, –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –Ω–∞ –µ–µ –∫–∞—á–µ—Å—Ç–≤–æ "–≤ —Ü–µ–ª–æ–º", *–Ω–∞ *–≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö –ø–æ—Ä–æ–≥–∞—Ö*.  –≠—Ç–æ –æ—Å–æ–±–µ–Ω–Ω–æ —Ü–µ–Ω–Ω–æ, –∫–æ–≥–¥–∞:

*   **–ù—É–∂–Ω–æ —Å—Ä–∞–≤–Ω–∏—Ç—å *—Ä–∞–∑–Ω—ã–µ –º–æ–¥–µ–ª–∏* –º–µ–∂–¥—É —Å–æ–±–æ–π *–±–µ–∑ –ø—Ä–∏–≤—è–∑–∫–∏ –∫ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–º—É –ø–æ—Ä–æ–≥—É*.**  –ö–∞–∫–∞—è –º–æ–¥–µ–ª—å *–≤ —Ü–µ–ª–æ–º* –ª—É—á—à–µ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä—É–µ—Ç?  ROC AUC –∏ PR AUC –¥–∞—é—Ç –æ—Ç–≤–µ—Ç –Ω–∞ —ç—Ç–æ—Ç –≤–æ–ø—Ä–æ—Å.
*   **–ù—É–∂–Ω–æ *–ø–æ–Ω—è—Ç—å, –∫–∞–∫ –∫–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–∏ –º–µ–Ω—è–µ—Ç—Å—è –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –ø–æ—Ä–æ–≥–∞*, –∏ *–Ω–∞–π—Ç–∏ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–π –ø–æ—Ä–æ–≥* –¥–ª—è –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–π –∑–∞–¥–∞—á–∏.**  ROC-–∫—Ä–∏–≤–∞—è –∏ PR-–∫—Ä–∏–≤–∞—è (–Ω–∞ –æ—Å–Ω–æ–≤–µ –∫–æ—Ç–æ—Ä—ã—Ö —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞—é—Ç—Å—è AUC) –ø–æ–∑–≤–æ–ª—è—é—Ç –≤–∏–∑—É–∞–ª—å–Ω–æ –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç—å *trade-off* –º–µ–∂–¥—É —Ä–∞–∑–Ω—ã–º–∏ –º–µ—Ç—Ä–∏–∫–∞–º–∏ (TPR –∏ FPR –¥–ª—è ROC, Precision –∏ Recall –¥–ª—è PR) –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ –ø–æ—Ä–æ–≥–∞.

#### **ROC AUC**

**1. ROC-–∫—Ä–∏–≤–∞—è ‚Äì  —ç—Ç–æ –≥—Ä–∞—Ñ–∏–∫, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É –¥–≤—É–º—è –≤–∞–∂–Ω—ã–º–∏ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞–º–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ –ø—Ä–∏ *–∏–∑–º–µ–Ω–µ–Ω–∏–∏ –ø–æ—Ä–æ–≥–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏*:**

*   **True Positive Rate (TPR) –∏–ª–∏ Recall (–ß—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å, –î–æ–ª—è –≤–µ—Ä–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤):**  TPR = TP / (TP + FN).  –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫—É—é –¥–æ–ª—é *—Ä–µ–∞–ª—å–Ω—ã—Ö* –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤ –º–æ–¥–µ–ª—å *–æ–±–Ω–∞—Ä—É–∂–∏–≤–∞–µ—Ç* (–∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä—É–µ—Ç –∫–∞–∫ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ) –ø—Ä–∏ –¥–∞–Ω–Ω–æ–º –ø–æ—Ä–æ–≥–µ.
*   **False Positive Rate (FPR) –∏–ª–∏ Fall-out (–î–æ–ª—è –ª–æ–∂–Ω—ã—Ö —Å—Ä–∞–±–∞—Ç—ã–≤–∞–Ω–∏–π, –î–æ–ª—è –ª–æ–∂–Ω–æ-–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤):**  FPR = FP / (FP + TN).  –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫—É—é –¥–æ–ª—é *—Ä–µ–∞–ª—å–Ω—ã—Ö* –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤ –º–æ–¥–µ–ª—å *–æ—à–∏–±–æ—á–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä—É–µ—Ç* –∫–∞–∫ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –ø—Ä–∏ –¥–∞–Ω–Ω–æ–º –ø–æ—Ä–æ–≥–µ.

**ROC-–∫—Ä–∏–≤–∞—è —Å—Ç—Ä–æ–∏—Ç—Å—è –≤ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö (FPR, TPR).**  –î–ª—è —Ä–∞–∑–Ω—ã—Ö –ø–æ—Ä–æ–≥–æ–≤ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –º—ã —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º FPR –∏ TPR, –∏ —Å—Ç—Ä–æ–∏–º —Ç–æ—á–∫–∏ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ.  –°–æ–µ–¥–∏–Ω—è—è —ç—Ç–∏ —Ç–æ—á–∫–∏, –ø–æ–ª—É—á–∞–µ–º ROC-–∫—Ä–∏–≤—É—é.



–ò–¥–µ–∞–ª—å–Ω—ã–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä –±—É–¥–µ—Ç —Å—Ç—Ä–µ–º–∏—Ç—å—Å—è –∫ *–≤–µ—Ä—Ö–Ω–µ–º—É –ª–µ–≤–æ–º—É —É–≥–ª—É* –≥—Ä–∞—Ñ–∏–∫–∞ ROC.  –ü–æ—á–µ–º—É?  –ü–æ—Ç–æ–º—É —á—Ç–æ –≤ –≤–µ—Ä—Ö–Ω–µ–º –ª–µ–≤–æ–º —É–≥–ª—É:

*   **TPR = 1 (100%) ‚Äì  –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω—ã–π True Positive Rate.**  –ú–æ–¥–µ–ª—å –æ–±–Ω–∞—Ä—É–∂–∏–≤–∞–µ—Ç **–≤—Å–µ** –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –æ–±—ä–µ–∫—Ç—ã.
*   **FPR = 0 (0%) ‚Äì  –º–∏–Ω–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω—ã–π False Positive Rate.**  –ú–æ–¥–µ–ª—å **–Ω–µ –¥–µ–ª–∞–µ—Ç *–Ω–∏ –æ–¥–Ω–æ–π* –ª–æ–∂–Ω–æ–π —Ç—Ä–µ–≤–æ–≥–∏.**

**–°–ª—É—á–∞–π–Ω—ã–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä (–∫–æ—Ç–æ—Ä—ã–π –ø—Ä–æ—Å—Ç–æ —Å–ª—É—á–∞–π–Ω–æ —É–≥–∞–¥—ã–≤–∞–µ—Ç –∫–ª–∞—Å—Å—ã) –±—É–¥–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω–æ–π –ª–∏–Ω–∏–∏ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ ROC (–ª–∏–Ω–∏—è "–±–µ–∑ —É–º–µ–Ω–∏—è"").**  –î–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ TPR ‚âà FPR –Ω–∞ –≤—Å–µ—Ö –ø–æ—Ä–æ–≥–∞—Ö.

**2. ROC AUC (Area Under the Curve - –ü–ª–æ—â–∞–¥—å –ø–æ–¥ ROC-–∫—Ä–∏–≤–æ–π)**

**ROC AUC ‚Äì  —ç—Ç–æ *–ø–ª–æ—â–∞–¥—å –ø–æ–¥ ROC-–∫—Ä–∏–≤–æ–π*.**  –≠—Ç–æ **—á–∏—Å–ª–µ–Ω–Ω–∞—è –º–µ—Ä–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞, –∫–æ—Ç–æ—Ä–∞—è *–∞–≥—Ä–µ–≥–∏—Ä—É–µ—Ç* –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é —Å–æ *–≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö –ø–æ—Ä–æ–≥–æ–≤* –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏.**

ROC AUC –º–æ–∂–Ω–æ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å –∫–∞–∫ *–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä *—Ä–∞–Ω–∂–∏—Ä—É–µ—Ç* —Å–ª—É—á–∞–π–Ω–æ –≤—ã–±—Ä–∞–Ω–Ω—ã–π *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π* –æ–±—ä–µ–∫—Ç *–≤—ã—à–µ*, —á–µ–º —Å–ª—É—á–∞–π–Ω–æ –≤—ã–±—Ä–∞–Ω–Ω—ã–π *–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π* –æ–±—ä–µ–∫—Ç.

–ß–µ–º *–±–æ–ª—å—à–µ* –ø–ª–æ—â–∞–¥—å –ø–æ–¥ ROC-–∫—Ä–∏–≤–æ–π (ROC AUC), —Ç–µ–º *–ª—É—á—à–µ* –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä.

*   **ROC AUC = 1:**  **–ò–¥–µ–∞–ª—å–Ω—ã–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä.**  –í—Å–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –æ–±—ä–µ–∫—Ç—ã —Ä–∞–Ω–∂–∏—Ä–æ–≤–∞–Ω—ã –≤—ã—à–µ –≤—Å–µ—Ö –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö.  ROC-–∫—Ä–∏–≤–∞—è –∏–¥–µ—Ç –≤ –≤–µ—Ä—Ö–Ω–∏–π –ª–µ–≤—ã–π —É–≥–æ–ª.
*   **ROC AUC = 0.5:**  **–°–ª—É—á–∞–π–Ω—ã–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä.**  –ù–µ—Ç —Ä–∞–∑–ª–∏—á–∏—è –º–µ–∂–¥—É —Ä–∞–Ω–∂–∏—Ä–æ–≤–∞–Ω–∏–µ–º –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤.  ROC-–∫—Ä–∏–≤–∞—è ‚Äì –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω–∞—è –ª–∏–Ω–∏—è.
*   **ROC AUC < 0.5:**  –ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä —Ä–∞–±–æ—Ç–∞–µ—Ç **—Ö—É–∂–µ, —á–µ–º —Å–ª—É—á–∞–π–Ω–æ–µ —É–≥–∞–¥—ã–≤–∞–Ω–∏–µ** (—Ö–æ—Ç—è —Ç–∞–∫–æ–µ —Ä–µ–¥–∫–æ –±—ã–≤–∞–µ—Ç –Ω–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ, –æ–±—ã—á–Ω–æ –º–æ–∂–Ω–æ "–ø–µ—Ä–µ–≤–µ—Ä–Ω—É—Ç—å" –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –∏ –ø–æ–ª—É—á–∏—Ç—å ROC AUC > 0.5).

ROC AUC ‚Äì  —É–¥–æ–±–Ω–∞—è –º–µ—Ç—Ä–∏–∫–∞ –¥–ª—è *—Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Ä–∞–∑–Ω—ã—Ö –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–æ–≤* –º–µ–∂–¥—É —Å–æ–±–æ–π.  –ß–µ–º –≤—ã—à–µ ROC AUC, —Ç–µ–º –ª—É—á—à–µ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä *–≤ —Ü–µ–ª–æ–º*.  ROC AUC *–º–µ–Ω–µ–µ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–∞* –∫ *–Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤*, —á–µ–º Accuracy, Precision, Recall –∏ F1-score.  –ü–æ—ç—Ç–æ–º—É ROC AUC —á–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É—é—Ç –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –º–æ–¥–µ–ª–µ–π –≤ –∑–∞–¥–∞—á–∞—Ö —Å –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –∫–ª–∞—Å—Å–∞–º–∏.

#### **PR AUC**

**PR AUC (Precision-Recall Area Under the Curve) ‚Äì  –∫–∞–∫ –∏ ROC AUC, –æ—Ü–µ–Ω–∏–≤–∞–µ—Ç –∫–∞—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ *–≤ —Ü–µ–ª–æ–º*, *–Ω–µ –∑–∞–≤–∏—Å—è –æ—Ç –ø–æ—Ä–æ–≥–∞*, –Ω–æ –¥–µ–ª–∞–µ—Ç —ç—Ç–æ **–Ω–µ–º–Ω–æ–≥–æ –ø–æ-–¥—Ä—É–≥–æ–º—É, –∏ *—Ñ–æ–∫—É—Å–∏—Ä—É–µ—Ç—Å—è –Ω–∞ *Precision* –∏ *Recall***, –∞ –Ω–µ –Ω–∞ TPR –∏ FPR, –∫–∞–∫ ROC AUC.**

**PR AUC ‚Äì  –æ—Å–æ–±–µ–Ω–Ω–æ –ø–æ–ª–µ–∑–Ω–∞ –≤ –∑–∞–¥–∞—á–∞—Ö —Å *—Å–∏–ª—å–Ω–æ –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –∫–ª–∞—Å—Å–∞–º–∏*, –∫–æ–≥–¥–∞ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å ‚Äì  —Ä–µ–¥–∫–∏–π, –Ω–æ *–æ—á–µ–Ω—å –≤–∞–∂–µ–Ω* –¥–ª—è –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è**, –∏ –∫–æ–≥–¥–∞ –Ω–∞—Å –±–æ–ª—å—à–µ –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç –∫–∞—á–µ—Å—Ç–≤–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∏–º–µ–Ω–Ω–æ –¥–ª—è *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞* (Precision –∏ Recall), –∞ –Ω–µ –æ–±—â–∞—è "—Ä–∞–∑–ª–∏—á–∏—Ç–µ–ª—å–Ω–∞—è —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å" –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ (–∫–∞–∫ –≤ ROC AUC).

**1. PR-–∫—Ä–∏–≤–∞—è (Precision-Recall curve)**

**PR-–∫—Ä–∏–≤–∞—è ‚Äì  —ç—Ç–æ –≥—Ä–∞—Ñ–∏–∫, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É *Precision* –∏ *Recall* –ø—Ä–∏ *–∏–∑–º–µ–Ω–µ–Ω–∏–∏ –ø–æ—Ä–æ–≥–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏*.**

PR-–∫—Ä–∏–≤–∞—è —Å—Ç—Ä–æ–∏—Ç—Å—è –≤ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö (Recall, Precision).  –î–ª—è —Ä–∞–∑–Ω—ã—Ö –ø–æ—Ä–æ–≥–æ–≤ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –º—ã —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º Precision –∏ Recall, –∏ —Å—Ç—Ä–æ–∏–º —Ç–æ—á–∫–∏ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ.  –°–æ–µ–¥–∏–Ω—è—è —ç—Ç–∏ —Ç–æ—á–∫–∏, –ø–æ–ª—É—á–∞–µ–º PR-–∫—Ä–∏–≤—É—é.



–ò–¥–µ–∞–ª—å–Ω—ã–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä –±—É–¥–µ—Ç —Å—Ç—Ä–µ–º–∏—Ç—å—Å—è –∫ *–≤–µ—Ä—Ö–Ω–µ–º—É –ø—Ä–∞–≤–æ–º—É —É–≥–ª—É* –≥—Ä–∞—Ñ–∏–∫–∞ PR.  –ü–æ—á–µ–º—É?  –ü–æ—Ç–æ–º—É —á—Ç–æ –≤ –≤–µ—Ä—Ö–Ω–µ–º –ø—Ä–∞–≤–æ–º —É–≥–ª—É:

*   **Precision = 1 (100%) ‚Äì  –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–∞—è Precision.**  –í—Å–µ –æ–±—ä–µ–∫—Ç—ã, –ø–æ–º–µ—á–µ–Ω–Ω—ã–µ –∫–∞–∫ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ, *–¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ* —è–≤–ª—è—é—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º–∏.  –ù–µ—Ç "–ª–æ–∂–Ω—ã—Ö —Ç—Ä–µ–≤–æ–≥" (FP = 0).
*   **Recall = 1 (100%) ‚Äì  –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω—ã–π Recall.**  –í—Å–µ *—Ä–µ–∞–ª—å–Ω—ã–µ* –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –æ–±—ä–µ–∫—Ç—ã *–æ–±–Ω–∞—Ä—É–∂–µ–Ω—ã* –∏ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω—ã –≤–µ—Ä–Ω–æ.  –ù–µ—Ç "–ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö —Ü–µ–ª–µ–π" (FN = 0).

**–°—Ä–∞–≤–Ω–µ–Ω–∏–µ ROC-–∫—Ä–∏–≤–æ–π –∏ PR-–∫—Ä–∏–≤–æ–π:**

*   **ROC-–∫—Ä–∏–≤–∞—è (Receiver Operating Characteristic):  (FPR, TPR)**.  –û—Å—å X ‚Äì FPR (False Positive Rate), –û—Å—å Y ‚Äì TPR (True Positive Rate).  –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç *trade-off* –º–µ–∂–¥—É –ª–æ–∂–Ω–æ-–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º–∏ —Å—Ä–∞–±–∞—Ç—ã–≤–∞–Ω–∏—è–º–∏ (FP) –∏ –≤–µ—Ä–Ω—ã–º–∏ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è–º–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤ (TP), –Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ *–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö* –∏ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö* –∫–ª–∞—Å—Å–æ–≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.  –ë–æ–ª—å—à–µ "—Å–º–æ—Ç—Ä–∏—Ç" –Ω–∞ –∫–∞—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ *–æ–±–æ–∏—Ö –∫–ª–∞—Å—Å–æ–≤* (–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ, –∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–≥–æ).

*   **PR-–∫—Ä–∏–≤–∞—è (Precision-Recall):  (Recall, Precision).**  –û—Å—å X ‚Äì Recall, –û—Å—å Y ‚Äì Precision.  –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç *trade-off* –º–µ–∂–¥—É —Ç–æ—á–Ω–æ—Å—Ç—å—é –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π (Precision) –∏ –ø–æ–ª–Ω–æ—Ç–æ–π –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤ (Recall).  –ë–æ–ª—å—à–µ "—Å–º–æ—Ç—Ä–∏—Ç" –Ω–∞ –∫–∞—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ *—Ç–æ–ª—å–∫–æ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞*, –∏ –∏–≥–Ω–æ—Ä–∏—Ä—É–µ—Ç *–∏—Å—Ç–∏–Ω–Ω–æ-–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ —Å—Ä–∞–±–∞—Ç—ã–≤–∞–Ω–∏—è (TN)*.

–í –∑–∞–¥–∞—á–∞—Ö —Å —Å–∏–ª—å–Ω–æ –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –∫–ª–∞—Å—Å–∞–º–∏, –∫–æ–≥–¥–∞ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å *–Ω–∞–º–Ω–æ–≥–æ –±–æ–ª—å—à–µ* –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ, ROC-–∫—Ä–∏–≤–∞—è –º–æ–∂–µ—Ç –¥–∞–≤–∞—Ç—å *–æ–ø—Ç–∏–º–∏—Å—Ç–∏—á–Ω—É—é* –æ—Ü–µ–Ω–∫—É –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏, –¥–∞–∂–µ –µ—Å–ª–∏ –º–æ–¥–µ–ª—å –ø–ª–æ—Ö–æ —Ä–∞–±–æ—Ç–∞–µ—Ç —Å *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º –∫–ª–∞—Å—Å–æ–º*.  –ü–æ—á–µ–º—É —Ç–∞–∫ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç?

**FPR (–æ—Å—å X –¥–ª—è ROC-–∫—Ä–∏–≤–æ–π) = FP / (FP + TN).**  –ï—Å–ª–∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å *–æ—á–µ–Ω—å –±–æ–ª—å—à–æ–π* (TN >> FP), —Ç–æ FPR –±—É–¥–µ—Ç *–æ—á–µ–Ω—å –º–∞–ª–µ–Ω—å–∫–∏–º*, –¥–∞–∂–µ –µ—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ FP –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –≤–µ–ª–∏–∫–æ –≤ *–∞–±—Å–æ–ª—é—Ç–Ω–æ–º –≤—ã—Ä–∞–∂–µ–Ω–∏–∏*.  ROC-–∫—Ä–∏–≤–∞—è –º–æ–∂–µ—Ç –≤—ã–≥–ª—è–¥–µ—Ç—å "—Ö–æ—Ä–æ—à–µ–π" (–±–ª–∏–∑–∫–æ–π –∫ –≤–µ—Ä—Ö–Ω–µ–º—É –ª–µ–≤–æ–º—É —É–≥–ª—É) –ø—Ä–æ—Å—Ç–æ –∑–∞ —Å—á–µ—Ç —Ç–æ–≥–æ, —á—Ç–æ FPR –æ—Å—Ç–∞–µ—Ç—Å—è –Ω–∏–∑–∫–∏–º –∏–∑-–∑–∞ –æ–≥—Ä–æ–º–Ω–æ–≥–æ TN, –¥–∞–∂–µ –µ—Å–ª–∏ TPR –ø—Ä–∏ —ç—Ç–æ–º –Ω–µ –æ—á–µ–Ω—å –≤—ã—Å–æ–∫–∏–π.

**PR-–∫—Ä–∏–≤–∞—è *–Ω–µ –∑–∞–≤–∏—Å–∏—Ç* –æ—Ç TN –∏ FPR, –∏ —Ñ–æ–∫—É—Å–∏—Ä—É–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –Ω–∞ TP, FP –∏ FN, —Ç–æ –µ—Å—Ç—å –Ω–∞ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–º –∫–ª–∞—Å—Å–µ –∏ –æ—à–∏–±–∫–∞—Ö, —Å–≤—è–∑–∞–Ω–Ω—ã—Ö —Å –Ω–∏–º*.**  Precision = TP / (TP + FP), Recall = TP / (TP + FN).  PR-–∫—Ä–∏–≤–∞—è *–±–æ–ª–µ–µ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–∞* –∫ –∫–∞—á–µ—Å—Ç–≤—É –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ *–∏–º–µ–Ω–Ω–æ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞*, –∏ –ª—É—á—à–µ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Ä–∞–∑–ª–∏—á–∏—è –º–µ–∂–¥—É –º–æ–¥–µ–ª—è–º–∏ –≤ —É—Å–ª–æ–≤–∏—è—Ö —Å–∏–ª—å–Ω–æ–≥–æ –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞ –∫–ª–∞—Å—Å–æ–≤, –æ—Å–æ–±–µ–Ω–Ω–æ –∫–æ–≥–¥–∞ –≤–∞–∂–Ω–∞ *—Ç–æ—á–Ω–æ—Å—Ç—å –∏ –ø–æ–ª–Ω–æ—Ç–∞ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞*.

**2. PR AUC (Area Under the Curve - –ü–ª–æ—â–∞–¥—å –ø–æ–¥ PR-–∫—Ä–∏–≤–æ–π)**

**PR AUC ‚Äì  —ç—Ç–æ *–ø–ª–æ—â–∞–¥—å –ø–æ–¥ PR-–∫—Ä–∏–≤–æ–π*.**  –≠—Ç–æ **—á–∏—Å–ª–µ–Ω–Ω–∞—è –º–µ—Ä–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞, –∫–æ—Ç–æ—Ä–∞—è *–∞–≥—Ä–µ–≥–∏—Ä—É–µ—Ç* –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é —Å *PR-–∫—Ä–∏–≤–æ–π* –∏ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç *—Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ Precision –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —É—Ä–æ–≤–Ω—è—Ö Recall*.**

PR AUC –º–æ–∂–Ω–æ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å –∫–∞–∫ *—Å—Ä–µ–¥–Ω–µ–≤–∑–≤–µ—à–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ Precision, –≤–∑–≤–µ—à–µ–Ω–Ω–æ–µ –ø–æ –∑–Ω–∞—á–µ–Ω–∏—è–º Recall*.  –ß–µ–º *–±–æ–ª—å—à–µ* –ø–ª–æ—â–∞–¥—å –ø–æ–¥ PR-–∫—Ä–∏–≤–æ–π (PR AUC), —Ç–µ–º *–ª—É—á—à–µ* –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä.

*   **PR AUC = 1:**  –ò–¥–µ–∞–ª—å–Ω—ã–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä (–≤ —Å–º—ã—Å–ª–µ Precision –∏ Recall).  PR-–∫—Ä–∏–≤–∞—è –∏–¥–µ—Ç –≤ –≤–µ—Ä—Ö–Ω–∏–π –ø—Ä–∞–≤—ã–π —É–≥–æ–ª.
*   **PR AUC —á–µ–º –≤—ã—à–µ, —Ç–µ–º –ª—É—á—à–µ.**  –ß–µ–º –±–æ–ª—å—à–µ PR AUC, —Ç–µ–º –ª—É—á—à–µ –±–∞–ª–∞–Ω—Å –º–µ–∂–¥—É Precision –∏ Recall, –∏ —Ç–µ–º –≤—ã—à–µ –∫–∞—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞*.
*   **PR AUC –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤.**  –î–ª—è —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤ PR AUC ‚âà 0.5, –Ω–æ –¥–ª—è *–Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö* –∫–ª–∞—Å—Å–æ–≤ PR AUC –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å *–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –Ω–∏–∂–µ 0.5*, –æ—Å–æ–±–µ–Ω–Ω–æ –µ—Å–ª–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å –æ—á–µ–Ω—å —Ä–µ–¥–∫–∏–π.  –ü–æ—ç—Ç–æ–º—É, –≤ –∑–∞–¥–∞—á–∞—Ö —Å —Ä–µ–¥–∫–∏–º–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º–∏ –∫–ª–∞—Å—Å–∞–º–∏, *–¥–∞–∂–µ –Ω–µ –æ—á–µ–Ω—å –≤—ã—Å–æ–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ PR AUC –º–æ–∂–µ—Ç –±—ã—Ç—å *–∑–Ω–∞—á–∏–º—ã–º —É–ª—É—á—à–µ–Ω–∏–µ–º* –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å–æ —Å–ª—É—á–∞–π–Ω—ã–º –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–æ–º*.

PR AUC ‚Äì  –æ—Å–æ–±–µ–Ω–Ω–æ —Ü–µ–Ω–Ω–∞ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –º–æ–¥–µ–ª–µ–π –≤ –∑–∞–¥–∞—á–∞—Ö —Å *—Å–∏–ª—å–Ω–æ –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –∫–ª–∞—Å—Å–∞–º–∏*, –∫–æ–≥–¥–∞ *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å ‚Äì  —Ä–µ–¥–∫–∏–π, –Ω–æ *–æ—á–µ–Ω—å –≤–∞–∂–µ–Ω* –¥–ª—è –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –∏ –∫–æ–≥–¥–∞ –≤–∞–∂–µ–Ω *–±–∞–ª–∞–Ω—Å –º–µ–∂–¥—É Precision –∏ Recall –∏–º–µ–Ω–Ω–æ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞*.  –ù–∞–ø—Ä–∏–º–µ—Ä, –≤ –∑–∞–¥–∞—á–∞—Ö –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –º–æ—à–µ–Ω–Ω–∏—á–µ—Å—Ç–≤–∞, —Ä–µ–¥–∫–∏—Ö –∑–∞–±–æ–ª–µ–≤–∞–Ω–∏–π, —Ç–∞—Ä–≥–µ—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —Ä–µ–∫–ª–∞–º—ã, –∫–æ–≥–¥–∞ –ø—Ä–æ–ø—É—Å—Ç–∏—Ç—å *–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π —Å–ª—É—á–∞–π* (–º–æ—à–µ–Ω–Ω–∏—á–µ—Å—Ç–≤–æ, –±–æ–ª–µ–∑–Ω—å, –∑–∞–∏–Ω—Ç–µ—Ä–µ—Å–æ–≤–∞–Ω–Ω–æ–≥–æ –∫–ª–∏–µ–Ω—Ç–∞) ‚Äì  –≥–æ—Ä–∞–∑–¥–æ —Ö—É–∂–µ, —á–µ–º –¥–æ–ø—É—Å—Ç–∏—Ç—å *–ª–æ–∂–Ω—É—é —Ç—Ä–µ–≤–æ–≥—É*.

---

#### **Confusion matrix –≤ python**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_classification # –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
from sklearn.model_selection import train_test_split # –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
from sklearn.linear_model import LogisticRegression # –ú–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ (–ø—Ä–æ—Å—Ç–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä)
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay # –§—É–Ω–∫—Ü–∏–∏ –¥–ª—è Confusion Matrix

# 1. –°–æ–∑–¥–∞–Ω–∏–µ —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ (2 –∫–ª–∞—Å—Å–∞)
X, y = make_classification(n_samples=1000, n_classes=2, weights=[0.9, 0.1], random_state=42) # –ù–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∫–ª–∞—Å—Å—ã (90% - –∫–ª–∞—Å—Å 0, 10% - –∫–ª–∞—Å—Å 1)

# 2. –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
model_logistic = LogisticRegression()
model_logistic.fit(X_train, y_train)

# 4. –ü–æ–ª—É—á–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
y_pred = model_logistic.predict(X_test)

# 5. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ Confusion Matrix
cm = confusion_matrix(y_test, y_pred) # –§—É–Ω–∫—Ü–∏—è confusion_matrix() –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∏—Å—Ç–∏–Ω–Ω—ã–µ –º–µ—Ç–∫–∏ (y_test) –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –º–µ—Ç–∫–∏ (y_pred)
print("Confusion Matrix (–º–∞—Å—Å–∏–≤ NumPy):\n", cm) # –í—ã–≤–æ–¥–∏–º Confusion Matrix –∫–∞–∫ –º–∞—Å—Å–∏–≤ NumPy

# 6. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è Confusion Matrix —Å –ø–æ–º–æ—â—å—é ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model_logistic.classes_) # –°–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏
disp.plot(cmap=plt.cm.Blues) # –°—Ç—Ä–æ–∏–º –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—é, cmap - —Ü–≤–µ—Ç–æ–≤–∞—è —Å—Ö–µ–º–∞ (—Å–∏–Ω—è—è)
plt.title('Confusion Matrix (–≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è)')
plt.show() # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –≥—Ä–∞—Ñ–∏–∫
```

* –ú–æ–¥–µ–ª—å —Ö–æ—Ä–æ—à–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä—É–µ—Ç –º–∞–∂–æ—Ä–∏—Ç–∞—Ä–Ω—ã–π –∫–ª–∞—Å—Å 0 (–≤—ã—Å–æ–∫–∏–π TN = 261, –Ω–∏–∑–∫–∏–π FP = 18). **Precision –¥–ª—è –∫–ª–∞—Å—Å–∞ 0** (—Ç–æ—á–Ω–æ—Å—Ç—å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ 0): TN / (TN + FN) = 261 / (261 + 9) ‚âà 96.6% - –æ—á–µ–Ω—å –≤—ã—Å–æ–∫–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å!

* –ú–æ–¥–µ–ª—å –ø–ª–æ—Ö–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä—É–µ—Ç –º–∏–Ω–æ—Ä–∏—Ç–∞—Ä–Ω—ã–π –∫–ª–∞—Å—Å 1 (–æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –Ω–∏–∑–∫–∏–π TP = 12, –≤—ã—Å–æ–∫–∏–π FN = 9, –Ω–æ –ø—Ä–∏ —ç—Ç–æ–º –¥–æ–≤–æ–ª—å–Ω–æ –≤—ã—Å–æ–∫–∏–π FP = 18). **Recall –¥–ª—è –∫–ª–∞—Å—Å–∞ 1** (–ø–æ–ª–Ω–æ—Ç–∞ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ 1): Recall = TP / (TP + FN) ‚âà 57% - –Ω–µ –æ—á–µ–Ω—å –≤—ã—Å–æ–∫–∏–π Recall. Precision –¥–ª—è –∫–ª–∞—Å—Å–∞ 1 (—Ç–æ—á–Ω–æ—Å—Ç—å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ 1): Precision = TP / (TP + FP) = 12 / (12 + 18) = 12 / 30 = 40% - —Ç–æ–∂–µ –Ω–µ–≤—ã—Å–æ–∫–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å.

* **F1-score –¥–ª—è –∫–ª–∞—Å—Å–∞ 1**: F1-score = 2 * (Precision * Recall) / (Precision + Recall) = 2*(0.4 * 0.57)/(0.4 + 0.57) = 0.456/0.97 $\approx$ 0.47 .

* **Accuracy** - –æ–±—â–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å –¥–ª—è –≤—Å–µ—Ö –∫–ª–∞—Å—Å–æ–≤: Accuracy = (TP + TN) / (TP + TN + FP + FN) = (12 + 261)/(12 + 261 + 18 + 9) = 273/300 = 0.91 .

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

precision = precision_score(y_test, y_pred) # Precision
recall = recall_score(y_test, y_pred) # Recall
f1 = f1_score(y_test, y_pred) # F1-score
accuracy = accuracy_score(y_test, y_pred) # Accuracy

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("Accuracy:", accuracy)
```

**1. –ù–µ–¥–æ–æ–±—É—á–µ–Ω–∏–µ (Underfitting) ‚Äì (High Bias, Low Variance)**

**–ù–µ–¥–æ–æ–±—É—á–µ–Ω–∏–µ (Underfitting) ‚Äì  —ç—Ç–æ —Å–∏—Ç—É–∞—Ü–∏—è, –∫–æ–≥–¥–∞ –º–æ–¥–µ–ª—å *—Å–ª–∏—à–∫–æ–º *–ø—Ä–æ—Å—Ç–∞—è* –∏ *–Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –≥–∏–±–∫–∞—è*, —á—Ç–æ–±—ã *—É–ª–æ–≤–∏—Ç—å *–∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–∏* –≤ *–æ–±—É—á–∞—é—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö***.**  –ú–æ–¥–µ–ª—å –Ω–µ "–¥–æ–Ω–∞—É—á–∏–ª–∞—Å—å", –Ω–µ "—Ä–∞–∑–æ–±—Ä–∞–ª–∞—Å—å" –≤ –¥–∞–Ω–Ω—ã—Ö, –∏ –ø–æ—ç—Ç–æ–º—É –ø–ª–æ—Ö–æ —Ä–∞–±–æ—Ç–∞–µ—Ç *–¥–∞–∂–µ –Ω–∞ *–æ–±—É—á–∞—é—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö*, –Ω–µ –≥–æ–≤–æ—Ä—è —É–∂–µ –æ –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö.

**–•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –Ω–µ–¥–æ–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏:**

*   **–ü–ª–æ—Ö–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ *–Ω–∞ *–æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ***.  –ú–µ—Ç—Ä–∏–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ (Accuracy, Precision, Recall, F1-score, MSE, MAE –∏ —Ç.–¥.) ‚Äì  –Ω–∏–∑–∫–∏–µ *–Ω–∞ *–æ–±—É—á–µ–Ω–∏–∏*.
*   **–ü–ª–æ—Ö–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ –∏ –Ω–∞ *—Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ***.  –ú–µ—Ç—Ä–∏–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ ‚Äì  –Ω–∏–∑–∫–∏–µ –∏ –Ω–∞ *—Ç–µ—Å—Ç–µ*.  –†–∞–∑—Ä—ã–≤ –º–µ–∂–¥—É –º–µ—Ç—Ä–∏–∫–∞–º–∏ –Ω–∞ –æ–±—É—á–µ–Ω–∏–∏ –∏ —Ç–µ—Å—Ç–µ ‚Äì  –Ω–µ–±–æ–ª—å—à–æ–π (–∏–ª–∏ –≤–æ–æ–±—â–µ –Ω–µ—Ç —Ä–∞–∑—Ä—ã–≤–∞).
*   **–ú–æ–¥–µ–ª—å *—Å–ª–∏—à–∫–æ–º –ø—Ä–æ—Å—Ç–∞—è***.  –ù–∞–ø—Ä–∏–º–µ—Ä, –ª–∏–Ω–µ–π–Ω–∞—è –º–æ–¥–µ–ª—å –¥–ª—è –Ω–µ–ª–∏–Ω–µ–π–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö, —Å–ª–∏—à–∫–æ–º "–º–µ–ª–∫–æ–µ" –¥–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π, –Ω–µ–π—Ä–æ—Å–µ—Ç—å —Å –º–∞–ª—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —Å–ª–æ–µ–≤ –∏ –Ω–µ–π—Ä–æ–Ω–æ–≤.
*   **–í—ã—Å–æ–∫–æ–µ —Å–º–µ—â–µ–Ω–∏–µ (Bias):**  –ú–æ–¥–µ–ª—å –¥–µ–ª–∞–µ—Ç —Å–∏–ª—å–Ω—ã–µ *—É–ø—Ä–æ—â–µ–Ω–∏—è* –∏ *–¥–æ–ø—É—â–µ–Ω–∏—è* –æ –¥–∞–Ω–Ω—ã—Ö, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç —Ä–µ–∞–ª—å–Ω–æ—Å—Ç–∏.  –°–º–µ—â–µ–Ω–∏–µ ‚Äì  —ç—Ç–æ —Å–∏—Å—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –æ—à–∏–±–∫–∞ –∏–∑-–∑–∞ *–Ω–µ–∞–¥–µ–∫–≤–∞—Ç–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏* –¥–ª—è –¥–∞–Ω–Ω—ã—Ö.  –ù–µ–¥–æ–æ–±—É—á–µ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å –∏–º–µ–µ—Ç –≤—ã—Å–æ–∫–æ–µ —Å–º–µ—â–µ–Ω–∏–µ (High Bias).
*   **–ù–∏–∑–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è (Variance):**  –ú–æ–¥–µ–ª—å –Ω–µ—á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–∞ –∫ –∏–∑–º–µ–Ω–µ–Ω–∏—è–º –≤ –æ–±—É—á–∞—é—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö, –µ–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è —Å—Ç–∞–±–∏–ª—å–Ω—ã**, –Ω–æ *–Ω–µ—Ç–æ—á–Ω—ã*.  –î–∏—Å–ø–µ—Ä—Å–∏—è ‚Äì  —ç—Ç–æ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ –∫ *—à—É–º—É* –∏ *–≤–∞—Ä–∏–∞—Ü–∏—è–º* –≤ –æ–±—É—á–∞—é—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö.  –ù–µ–¥–æ–æ–±—É—á–µ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å –∏–º–µ–µ—Ç –Ω–∏–∑–∫—É—é –¥–∏—Å–ø–µ—Ä—Å–∏—é (Low Variance).

**2. –ü–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ (Overfitting) ‚Äì (High Variance, Low Bias)**

**–ü–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ (Overfitting) ‚Äì  —ç—Ç–æ —Å–∏—Ç—É–∞—Ü–∏—è, –∫–æ–≥–¥–∞ –º–æ–¥–µ–ª—å *—Å–ª–∏—à–∫–æ–º *—Å–ª–æ–∂–Ω–∞—è* –∏ *—Å–ª–∏—à–∫–æ–º "–ø–æ–¥—Å—Ç—Ä–æ–∏–ª–∞—Å—å"* –ø–æ–¥ *–∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã–µ *–æ–±—É—á–∞—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ***, –≤–∫–ª—é—á–∞—è **—à—É–º –∏ —Å–ª—É—á–∞–π–Ω—ã–µ –∫–æ–ª–µ–±–∞–Ω–∏—è** –≤ —ç—Ç–∏—Ö –¥–∞–Ω–Ω—ã—Ö.  –ú–æ–¥–µ–ª—å "–∑–∞–ø–æ–º–Ω–∏–ª–∞" –æ–±—É—á–∞—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ "–Ω–∞–∏–∑—É—Å—Ç—å", –Ω–æ –ø–æ—Ç–µ—Ä—è–ª–∞ —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å –∫ *–æ–±–æ–±—â–µ–Ω–∏—é* –Ω–∞ –Ω–æ–≤—ã–µ, *–Ω–µ–≤–∏–¥–∞–Ω–Ω—ã–µ —Ä–∞–Ω–µ–µ* –¥–∞–Ω–Ω—ã–µ.

**–•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏:**

*   **–û—Ç–ª–∏—á–Ω–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ *–Ω–∞ *–æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ***.  –ú–µ—Ç—Ä–∏–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ ‚Äì  –æ—á–µ–Ω—å –≤—ã—Å–æ–∫–∏–µ *–Ω–∞ *–æ–±—É—á–µ–Ω–∏–∏* (—á–∞—Å—Ç–æ –ø–æ—á—Ç–∏ 100% Accuracy, MSE ‚âà 0 –∏ —Ç.–¥.).
*   **–ü–ª–æ—Ö–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ –Ω–∞ *—Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ***.  –ú–µ—Ç—Ä–∏–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ ‚Äì  –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ *—Ö—É–∂–µ* –Ω–∞ *—Ç–µ—Å—Ç–µ*, —á–µ–º –Ω–∞ –æ–±—É—á–µ–Ω–∏–∏.  –ë–æ–ª—å—à–æ–π —Ä–∞–∑—Ä—ã–≤ –º–µ–∂–¥—É –º–µ—Ç—Ä–∏–∫–∞–º–∏ –Ω–∞ –æ–±—É—á–µ–Ω–∏–∏ –∏ —Ç–µ—Å—Ç–µ ‚Äì  *–≥–ª–∞–≤–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫ –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏—è* (Overfitting Gap).
*   **–ú–æ–¥–µ–ª—å —Å–ª–∏—à–∫–æ–º *—Å–ª–æ–∂–Ω–∞—è**.  –ù–∞–ø—Ä–∏–º–µ—Ä, —Å–ª–∏—à–∫–æ–º "–≥–ª—É–±–æ–∫–æ–µ" –¥–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π, –Ω–µ–π—Ä–æ—Å–µ—Ç—å —Å –æ–≥—Ä–æ–º–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —Å–ª–æ–µ–≤ –∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.
*   **–ù–∏–∑–∫–æ–µ —Å–º–µ—â–µ–Ω–∏–µ (Bias):**  –ú–æ–¥–µ–ª—å —Ö–æ—Ä–æ—à–æ *–ø—Ä–∏–±–ª–∏–∂–∞–µ—Ç* –æ–±—É—á–∞—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ**, –ø–æ—á—Ç–∏ –Ω–µ –¥–µ–ª–∞–µ—Ç —Å–∏—Å—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö –æ—à–∏–±–æ–∫ –Ω–∞ –æ–±—É—á–µ–Ω–∏–∏.  –°–º–µ—â–µ–Ω–∏–µ ‚Äì  –Ω–∏–∑–∫–æ–µ (Low Bias).
*   **–í—ã—Å–æ–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è (Variance):**  –ú–æ–¥–µ–ª—å –æ—á–µ–Ω—å *—á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–∞* –∫ *–º–∞–ª–µ–π—à–∏–º –∏–∑–º–µ–Ω–µ–Ω–∏—è–º* –≤ –æ–±—É—á–∞—é—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö, –µ–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –æ—á–µ–Ω—å —Å–∏–ª—å–Ω–æ –º–µ–Ω—è—é—Ç—Å—è –ø—Ä–∏ –Ω–µ–±–æ–ª—å—à–∏—Ö –∏–∑–º–µ–Ω–µ–Ω–∏—è—Ö –≤ –¥–∞–Ω–Ω—ã—Ö.  –î–∏—Å–ø–µ—Ä—Å–∏—è ‚Äì  –≤—ã—Å–æ–∫–∞—è (High Variance). –ü–µ—Ä–µ–æ–±—É—á–µ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å "—Ä–µ–∞–≥–∏—Ä—É–µ—Ç –Ω–∞ –∫–∞–∂–¥—ã–π "—á–∏—Ö" –≤ –¥–∞–Ω–Ω—ã—Ö, –≤–∫–ª—é—á–∞—è —à—É–º –∏ —Å–ª—É—á–∞–π–Ω–æ—Å—Ç—å.

**3. Bias-Variance Tradeoff (–ö–æ–º–ø—Ä–æ–º–∏—Å—Å –°–º–µ—â–µ–Ω–∏—è –∏ –î–∏—Å–ø–µ—Ä—Å–∏–∏)**

**Bias-Variance Tradeoff ‚Äì  —ç—Ç–æ *—Ñ—É–Ω–¥–∞–º–µ–Ω—Ç–∞–ª—å–Ω–∞—è –ø—Ä–æ–±–ª–µ–º–∞* –≤ –º–∞—à–∏–Ω–Ω–æ–º –æ–±—É—á–µ–Ω–∏–∏.**  –ù–µ–≤–æ–∑–º–æ–∂–Ω–æ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –∏–º–µ—Ç—å –º–æ–¥–µ–ª—å —Å *–Ω–∏–∑–∫–∏–º —Å–º–µ—â–µ–Ω–∏–µ–º* (—Ö–æ—Ä–æ—à–æ "–ø–æ–¥–≥–æ–Ω—è–µ—Ç—Å—è" –ø–æ–¥ –æ–±—É—á–∞—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ) –∏ *–Ω–∏–∑–∫–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π* (—Ö–æ—Ä–æ—à–æ "–æ–±–æ–±—â–∞–µ—Ç" –Ω–∞ –Ω–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ).  **–£–º–µ–Ω—å—à–∞—è —Å–º–µ—â–µ–Ω–∏–µ, –º—ã –æ–±—ã—á–Ω–æ *—É–≤–µ–ª–∏—á–∏–≤–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é*, –∏ –Ω–∞–æ–±–æ—Ä–æ—Ç.  –≠—Ç–æ ‚Äì  –∫–∞–∫ **–∫–∞—á–µ–ª–∏.

–ò–¥–µ–∞–ª—å–Ω–∞—è –º–æ–¥–µ–ª—å ‚Äì  —ç—Ç–æ —Ç–∞, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç *–±–∞–ª–∞–Ω—Å* –º–µ–∂–¥—É —Å–º–µ—â–µ–Ω–∏–µ–º –∏ –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π, –∏ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç *–æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–≥–æ –∫–∞—á–µ—Å—Ç–≤–∞* –Ω–∞ *–Ω–æ–≤—ã—Ö, –Ω–µ–≤–∏–¥–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö*.

**–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è Bias-Variance Tradeoff:**



Bias-Variance Tradeoff —á–∞—Å—Ç–æ –∏–∑–æ–±—Ä–∞–∂–∞—é—Ç –≤ –≤–∏–¥–µ **U-–æ–±—Ä–∞–∑–Ω–æ–π –∫—Ä–∏–≤–æ–π**, –≥–¥–µ –æ—Å—å X ‚Äì  —Å–ª–æ–∂–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏, –æ—Å—å Y ‚Äì  –æ—à–∏–±–∫–∞ –º–æ–¥–µ–ª–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, MSE, MAE).

*   **–°–ª–µ–≤–∞ (–Ω–∏–∑–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏):  –ù–µ–¥–æ–æ–±—É—á–µ–Ω–∏–µ (Underfitting).**  –ú–æ–¥–µ–ª—å *—Å–ª–∏—à–∫–æ–º –ø—Ä–æ—Å—Ç–∞—è*, –≤—ã—Å–æ–∫–æ–µ —Å–º–µ—â–µ–Ω–∏–µ (High Bias), –≤—ã—Å–æ–∫–∞—è –æ—à–∏–±–∫–∞ *–Ω–∞ *–æ–±—É—á–µ–Ω–∏–∏ –∏ —Ç–µ—Å—Ç–µ*.
*   **–°–ø—Ä–∞–≤–∞ (–≤—ã—Å–æ–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏):  –ü–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ (Overfitting).**  –ú–æ–¥–µ–ª—å *—Å–ª–∏—à–∫–æ–º —Å–ª–æ–∂–Ω–∞—è*, –≤—ã—Å–æ–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è (High Variance), –Ω–∏–∑–∫–∞—è –æ—à–∏–±–∫–∞ *–Ω–∞ *–æ–±—É—á–µ–Ω–∏–∏*, –Ω–æ –≤—ã—Å–æ–∫–∞—è –æ—à–∏–±–∫–∞ *–Ω–∞ *—Ç–µ—Å—Ç–µ* (–±–æ–ª—å—à–æ–π —Ä–∞–∑—Ä—ã–≤ –º–µ–∂–¥—É –æ–±—É—á–µ–Ω–∏–µ–º –∏ —Ç–µ—Å—Ç–æ–º).
*   **–í —Ü–µ–Ω—Ç—Ä–µ (–æ–ø—Ç–∏–º–∞–ª—å–Ω–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏):  "Sweet Spot" (–ó–æ–ª–æ—Ç–∞—è —Å–µ—Ä–µ–¥–∏–Ω–∞).**  –ú–æ–¥–µ–ª—å –∏–º–µ–µ—Ç –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–π –±–∞–ª–∞–Ω—Å —Å–º–µ—â–µ–Ω–∏—è –∏ –¥–∏—Å–ø–µ—Ä—Å–∏–∏, –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –æ—à–∏–±–∫—É –Ω–∞ *—Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ*.

**–ú–µ—Ç–æ–¥—ã –±–æ—Ä—å–±—ã —Å –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ–º (High Variance):**

*   **–†–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è (Regularization):**  –î–æ–±–∞–≤–ª–µ–Ω–∏–µ "—à—Ç—Ä–∞—Ñ–∞" –∫ —Ñ—É–Ω–∫—Ü–∏–∏ –æ—à–∏–±–∫–∏ –∑–∞ **—Å–ª–∏—à–∫–æ–º –±–æ–ª—å—à–∏–µ –≤–µ—Å–∞ –º–æ–¥–µ–ª–∏**.  –£–ø—Ä–æ—â–∞–µ—Ç –º–æ–¥–µ–ª—å, —É–º–µ–Ω—å—à–∞–µ—Ç –¥–∏—Å–ø–µ—Ä—Å–∏—é. *   **–ö—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è (Cross-validation):**  –û—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏ –Ω–∞ *–Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö —Ä–∞–∑–Ω—ã—Ö –ø–æ–¥–≤—ã–±–æ—Ä–∫–∞—Ö* –¥–∞–Ω–Ω—ã—Ö.  –ü–æ–º–æ–≥–∞–µ—Ç **–±–æ–ª–µ–µ –Ω–∞–¥–µ–∂–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å –æ–±–æ–±—â–∞—é—â—É—é —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏** –∏ **–≤—ã–±—Ä–∞—Ç—å –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã**.
*   **–£–º–µ–Ω—å—à–µ–Ω–∏–µ —Å–ª–æ–∂–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏:**  –£–ø—Ä–æ—â–µ–Ω–∏–µ –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—ã –º–æ–¥–µ–ª–∏, —É–º–µ–Ω—å—à–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ (–Ω–∞–ø—Ä–∏–º–µ—Ä, —É–º–µ–Ω—å—à–µ–Ω–∏–µ –≥–ª—É–±–∏–Ω—ã –¥–µ—Ä–µ–≤–∞, –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–ª–æ–µ–≤ –∏ –Ω–µ–π—Ä–æ–Ω–æ–≤ –≤ –Ω–µ–π—Ä–æ—Å–µ—Ç–∏).
*   **–†–∞–Ω–Ω—è—è –æ—Å—Ç–∞–Ω–æ–≤–∫–∞ (Early stopping):**  –û—Å—Ç–∞–Ω–æ–≤–∫–∞ –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏ **–¥–æ —Ç–æ–≥–æ, –∫–∞–∫ –æ–Ω–∞ "–ø–µ—Ä–µ–æ–±—É—á–∏—Ç—Å—è"**.  –ú–æ–Ω–∏—Ç–æ—Ä–∏–Ω–≥ –æ—à–∏–±–∫–∏ –Ω–∞ *–≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ* –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ –æ–±—É—á–µ–Ω–∏—è, –∏ –æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –æ–±—É—á–µ–Ω–∏—è, –∫–æ–≥–¥–∞ –æ—à–∏–±–∫–∞ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–∏ –Ω–∞—á–∏–Ω–∞–µ—Ç —Ä–∞—Å—Ç–∏ (–ø—Ä–∏–∑–Ω–∞–∫ –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏—è).

(PCA, t-SNE, Feature Importance ‚Äì  –±—É–¥–µ–º –∏–∑—É—á–∞—Ç—å –ø–æ–∑–∂–µ).
*   **–ê–Ω—Å–∞–º–±–ª–µ–≤—ã–µ –º–µ—Ç–æ–¥—ã (Ensemble methods):**  **–ö–æ–º–±–∏–Ω–∞—Ü–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π *–Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö *—Å–ª–∞–±—ã—Ö* –º–æ–¥–µ–ª–µ–π* –≤ –æ–¥–Ω—É *—Å–∏–ª—å–Ω—É—é* –º–æ–¥–µ–ª—å.**  –ê–Ω—Å–∞–º–±–ª–∏ —á–∞—Å—Ç–æ –ø–æ–º–æ–≥–∞—é—Ç **—É–º–µ–Ω—å—à–∏—Ç—å –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —É–ª—É—á—à–∏—Ç—å –æ–±–æ–±—â–∞—é—â—É—é —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å**.  (Random Forest, Gradient Boosting ‚Äì  –±—É–¥–µ–º –∏–∑—É—á–∞—Ç—å –ø–æ–∑–∂–µ).
*   **–ë–æ–ª—å—à–µ –¥–∞–Ω–Ω—ã—Ö (More data):**  –í –∏–¥–µ–∞–ª–µ, **—É–≤–µ–ª–∏—á–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏** ‚Äì  –æ–¥–∏–Ω –∏–∑ —Å–∞–º—ã—Ö –Ω–∞–¥–µ–∂–Ω—ã—Ö —Å–ø–æ—Å–æ–±–æ–≤ **—É–º–µ–Ω—å—à–∏—Ç—å –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ**.  –ß–µ–º –±–æ–ª—å—à–µ –¥–∞–Ω–Ω—ã—Ö, —Ç–µ–º –ª—É—á—à–µ –º–æ–¥–µ–ª—å —É—á–∏—Ç—Å—è "–≤–∏–¥–µ—Ç—å –æ–±—â–∏–µ –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–∏", –∞ –Ω–µ "–∑–∞–ø–æ–º–∏–Ω–∞—Ç—å" —à—É–º –∏ —Å–ª—É—á–∞–π–Ω–æ—Å—Ç–∏.
