
### III. Визуализация данных

Визуализация — ключевой инструмент для понимания данных, выявления закономерностей и представления результатов.

#### III.A. Matplotlib: Основы Построения Графиков

**Matplotlib** — фундаментальная библиотека Python для создания статических, анимированных и интерактивных визуализаций. `matplotlib.pyplot` — основной интерфейс для быстрого построения графиков.

import matplotlib.pyplot as plt
import numpy as np

**1. Простой Линейный График (`plt.plot()`):**
Показывает зависимость Y от X в виде линии.

```python
import matplotlib.pyplot as plt
import numpy as np

x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 1, 3, 5]

plt.plot(x_values, y_values) # Основная функция
plt.show() # Отображает график
```

**2. Настройка Элементов Графика:**
*   **Заголовок:** `plt.title('Заголовок')`
*   **Подписи осей:** `plt.xlabel('Подпись X')`, `plt.ylabel('Подпись Y')`
*   **Цвет и стиль линии** (в `plt.plot()`):
    *   `color='цвет'` (e.g., `'red'`, `'blue'`, `'#FF5733'`)
    *   `linestyle='стиль'` или `ls='стиль'` (`'-'`, `'--'`, `':'`, `'-.'`)
    *   `linewidth=толщина`
*   **Маркеры точек:** `marker='стиль_маркера'` (`'o'`, `'x'`, `'s'`, `'+'`, ...). Полезно, если линия не нужна (`ls='None'`).

```python
plt.plot(x_values, y_values, color='green', linestyle='--', marker='o', linewidth=2)
plt.title('Настроенный график')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True) # Добавить сетку
plt.legend(['Линия Y']) # Добавить легенду (если несколько линий)
plt.show()
```

**3. Точечные Диаграммы (`plt.scatter()`):**
Показывает отдельные точки данных, используется для исследования взаимосвязи, выявления кластеров и выбросов.

```python
plt.scatter(x_values, y_values)
plt.title('Точечная диаграмма')
plt.show()
```

*   **Настройка точек** (в `plt.scatter()`):
    *   `marker='стиль'` (см. выше)
    *   `s=размер` (размер точки в пунктах^2)
    *   `c=цвет` или `color=цвет`
    *   `alpha=прозрачность` (от 0 до 1)


```python
plt.scatter(x_values, y_values, color='purple', marker='x', s=100, alpha=0.7)
plt.show()
```

*4. Столбчатые Диаграммы (`plt.bar()`):**
Сравнивает числовые значения между **дискретными категориями**.

```python
categories = ['A', 'B', 'C', 'D']
values = [15, 22, 18, 25]

plt.bar(categories, values)
plt.title('Столбчатая диаграмма')
plt.show()
```

*   **Настройка столбцов** (в `plt.bar()`):
    *   `color='цвет'`
    *   `width=ширина` (0 до 1, default 0.8)
    *   `edgecolor=цвет_границы`
    *   `linewidth=толщина_границы`


```python
plt.bar(categories, values, color='gray', edgecolor='black', width=0.6)
plt.show()
```

**5. Гистограммы (`plt.hist()`):**
Показывает **распределение** *непрерывных* данных. Ось X - интервалы значений (bins), Ось Y - частота попадания в интервал.

```python
data = np.random.randn(1000) # Пример данных

plt.hist(data)
plt.title('Гистограмма')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.show()
```

*   **Настройка гистограммы** (в `plt.hist()`):
    *   `bins=число_или_границы` (количество интервалов или их границы)
    *   `color='цвет'`
    *   `edgecolor=цвет_границы`
    *   `alpha=прозрачность`
    *   `histtype='тип'` (`'bar'`, `'step'`, `'stepfilled'`)

```python
plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.8)
plt.show()
plt.close()
```

#### III.B. Seaborn: Статистическая Визуализация

**Seaborn** — библиотека для создания информативных и привлекательных статистических графиков, построенная поверх Matplotlib.

**Преимущества:**
*   Высокоуровневый интерфейс.
*   Специализация на статистических графиках.
*   **Отличная интеграция с Pandas DataFrame.**
*   Привлекательный внешний вид по умолчанию.
*   Расширяет Matplotlib (можно использовать функции `plt` для донастройки).

```python
import seaborn as sns
import pandas as pd
```


**1. Точечные Диаграммы (`sns.scatterplot()`):**
Аналог `plt.scatter`, но удобнее работать с DataFrame.

```python
# Пример DataFrame
data_dict = {
    'Возраст': [25, 30, 28, 35, 22, 40, 29, 33, 45, 27],
    'Доход': [30, 50, 45, 60, 25, 70, 48, 55, 80, 40], # в тыс.
    'Пол': ['Жен', 'Муж', 'Муж', 'Муж', 'Жен', 'Муж', 'Жен', 'Жен', 'Муж', 'Жен'],
    'Стаж': [2, 7, 5, 10, 1, 15, 6, 8, 20, 4]
}
df = pd.DataFrame(data_dict)

# Базовый scatterplot
sns.scatterplot(data=df, x='Возраст', y='Доход')
plt.title('Доход от Возраста')
plt.show()
plt.close()
```

**2. Кодирование Дополнительных Измерений:**
Использование визуальных атрибутов (цвет, размер, стиль) для отображения других переменных.
*   **`hue='категориальный_столбец'`:** Цвет точек зависит от категории. Seaborn автоматически добавляет легенду.
*   **`size='числовой_столбец'`:** Размер точек зависит от числового значения.
*   **`style='категориальный_столбец'`:** Форма маркера зависит от категории.

```python
# Scatterplot с hue
sns.scatterplot(data=df, x='Возраст', y='Доход', hue='Пол')
plt.title('Доход от Возраста (по полу)')
plt.show()
plt.close()

# Scatterplot с size
sns.scatterplot(data=df, x='Возраст', y='Доход', size='Стаж')
plt.title('Доход от Возраста (размер по стажу)')
plt.show()
plt.close()

# Scatterplot с hue и style
sns.scatterplot(data=df, x='Возраст', y='Доход', hue='Пол', style='Пол') # Можно использовать одну и ту же переменную
plt.title('Доход от Возраста (цвет и стиль по полу)')
plt.show()
plt.close()
```

**3. Ящичковые Диаграммы (`sns.boxplot()`):**
Сравнивает **распределения** числовой переменной по категориям. Показывает медиану, квартили (Q1, Q3), межквартильный размах (IQR), "усы" (обычно Q1/Q3 ± 1.5*IQR) и выбросы.

```python
sns.boxplot(data=df, x='Пол', y='Доход', hue='Пол') # hue здесь для цвета ящиков
plt.title('Распределение Дохода по Полу (Box Plot)')
plt.show()
plt.close()
```

*   **`hue`:** Позволяет добавить еще одно категориальное разделение.
*   **`palette`:** Настройка цветовой палитры.

**4. Гистограммы (`sns.histplot()`):**
Аналог `plt.hist`, удобнее с DataFrame. Показывает распределение непрерывной переменной.



```python
sns.histplot(data=df, x='Возраст', bins=5, color='skyblue')
plt.title('Распределение Возраста')
plt.show()
plt.close()
```

*   **`kde=True`:** Добавляет сглаженную кривую оценки плотности ядра (Kernel Density Estimate).
*   **`hue`:** Позволяет строить гистограммы для разных категорий на одном графике (часто с некоторой прозрачностью `alpha`).

```python
sns.histplot(data=df, x='Доход', hue='Пол', kde=True, bins=6)
plt.title('Распределение Дохода по Полу (Hist + KDE)')
plt.show()
plt.close()
```

**5. Скрипичные Диаграммы (`sns.violinplot()`):**
Комбинация **boxplot** и **KDE plot**. Показывает форму распределения (плотность) по бокам и сводную статистику (часто boxplot) внутри.

```python
sns.violinplot(data=df, x='Пол', y='Доход', hue='Пол')
plt.title('Распределение Дохода по Полу (Violin Plot)')
plt.show()
plt.close()
```

*   **`inner='тип'`:** `'box'`, `'quartile'`, `'point'`, `'stick'`, `None` - настройка внутреннего отображения.
*   **`split=True`:** (При использовании `hue` с 2 категориями) рисует "половинки" скрипок для каждой категории для более компактного сравнения.

```python
# Пример с split=True (если бы была еще одна категория для hue)
# sns.violinplot(data=df, x='Category', y='Value', hue='BinaryVar', split=True)
# plt.show()
# plt.close()
```

**6. Парные Диаграммы (`sns.pairplot()`):**
Матрица графиков для визуализации **попарных взаимосвязей** между всеми **числовыми** переменными в DataFrame.
*   **Диагональ:** Гистограммы или KDE для каждой переменной.
*   **Вне диагонали:** Точечные диаграммы (scatter plots) для каждой пары переменных.

```python
# Строим pairplot для числовых столбцов, кодируя цветом по 'Пол'
sns.pairplot(df, hue='Пол', vars=['Возраст', 'Доход', 'Стаж']) # Указываем числовые переменные в vars
plt.suptitle('Парные диаграммы (Возраст, Доход, Стаж) с разделением по полу', y=1.02) # Добавляем общий заголовок
plt.show()
plt.close()
```

*   **`vars=[...]`:** Список столбцов для включения в график.
*   **`kind='reg'`:** Нарисовать линии регрессии на scatter plot'ах.
*   **`diag_kind='kde'`:** Использовать KDE на диагонали вместо гистограмм.

**7. Тепловые Карты (`sns.heatmap()`):**
Визуализация матрицы данных, где значения кодируются цветом. Часто используется для **корреляционных матриц**.

```python
# Выбираем только числовые столбцы для корреляции
numeric_df = df[['Возраст', 'Доход', 'Стаж']]
correlation_matrix = numeric_df.corr() # Метод Pandas для расчета корреляций

print("Корреляционная матрица:")
print(correlation_matrix)

plt.figure(figsize=(7, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
# annot=True: показывать значения в ячейках
# cmap='coolwarm': расходящаяся цветовая палитра (синий-белый-красный)
# fmt=".2f": формат чисел (2 знака после запятой)
# linewidths: линии между ячейками
plt.title('Тепловая карта корреляций')
plt.show()
plt.close()
```
