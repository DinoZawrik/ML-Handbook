## Линейная Регрессия
#### **Уравнение прямой**

Уравнение прямой на плоскости выглядит так:

`y = wx + b`

Где:

*   `y` - это **зависимая переменная** (то, что мы хотим предсказать).  В машинном обучении ее часто называют **целевой переменной** или **таргетом**.
*   `x` - это **независимая переменная** (то, от чего зависит `y`). В машинном обучении это называется **признаком** или **фичей**.
*   `w` - это **наклон прямой** (или **вес**, weight). Он показывает, насколько сильно `y` меняется при изменении `x` на единицу. Если `w` положительный, то при увеличении `x` значение `y` тоже увеличивается.
*   `b` - это **смещение** (или **свободный член**, bias). Это значение `y`, когда `x` равен нулю.

**В машинном обучении `w` и `b` называются параметрами модели.**  Наша задача - **найти такие значения `w` и `b`, чтобы прямая `y = wx + b` наилучшим образом описывала наши данные.**

**Для случая с несколькими признаками (множественная линейная регрессия)** уравнение будет немного сложнее, но принцип тот же:

$y = w₁x₁ + w₂x₂ + ... + wₙxₙ + b$

Где:

*   $x₁, x₂, ..., xₙ$ - это **признаки** (например, площадь, количество комнат, расстояние до метро и т.д.).
*   $w₁, w₂, ..., wₙ$ - это **веса** для каждого признака. Они показывают, насколько важен каждый признак для предсказания `y`.

#### **Функция ошибки (Mean Squared Error - MSE)**

Как оценить качество прямой `y = wx + b`?

Нужна **мера "плохости"** прямой - **функция ошибки**.
**MSE (Mean Squared Error) - Среднеквадратичная ошибка** - распространенная функция ошибки для регрессии. **MSE:** мера "плохости" прямой, показывает средний квадрат ошибок предсказаний. Чем меньше MSE, тем лучше прямая.

**Вот как она работает:**

1.  Для каждой **точки данных** в наших данных мы берем ее **фактическое значение** (из таблицы) и **предсказанное значение** (которое мы получаем, используя нашу модель).
2.  Находим **разницу** между фактическим и предсказанным значением. Эта разница называется **ошибкой** или **отклонением**.
3.  Возводим эту ошибку **в квадрат**.  Это делается для того, чтобы:
    *   Избавиться от знака ошибки (чтобы и положительные, и отрицательные ошибки вносили вклад в общую "плохость").
    *   Увеличить "штраф" за большие ошибки. Квадрат больших чисел растет быстрее, чем квадрат маленьких.
4.  Суммируем **квадраты ошибок** для **всех точек данных** в наших данных.
5.  Делим полученную сумму на **количество точек данных**, чтобы получить **среднее значение** квадратов ошибок.  Отсюда и название - **Среднеквадратичная ошибка (MSE)**.

**Формула MSE выглядит так:**

$MSE = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$

Где:

*   `N` - количество домов в данных.
*   `yᵢ` - фактическая цена `i`-го дома.
*   `ŷᵢ` - предсказанная цена `i`-го дома (вычисленная по формуле `ŷᵢ = wxᵢ + b`, где `xᵢ` - площадь `i`-го дома).
*   `Σ` - знак суммы (суммирование по всем домам от 1 до N).

**Чем меньше значение MSE, тем лучше наша прямая "подходит" под данные.**  Идеально, если MSE равно 0, это означало бы, что все предсказанные цены идеально совпадают с фактическими (что в реальной жизни почти никогда не бывает).

**Наша цель в линейной регрессии - найти такие значения параметров `w` и `b`, которые минимизируют значение MSE.**  То есть, мы хотим подобрать такую прямую, чтобы **средний квадрат отклонений предсказанных цен от фактических был как можно меньше.**


**как же найти те самые "лучшие" значения параметров `w` и `b`, которые минимизируют нашу Среднеквадратичную ошибку (СКО).**

Существует **два основных подхода**:

1. **МНК (Метод наименьших квадратов):**
    - **Тип:** Аналитический.
    - **Суть:** Находит `w` и `b` **сразу** по формулам.
    - **Применение:** Линейная регрессия и ее расширения.
    - **Преимущество:** Точное решение для линейных моделей.

2. **Градиентный спуск:**
    - **Тип:** Итеративный.
    - **Суть:** Подбирает `w` и `b` **постепенно**, шаг за шагом.
    - **Применение:** Широкий спектр моделей, включая сложные (нейросети).
    - **Преимущество:** Универсальность, подходит для нелинейных моделей.

**3. Метод наименьших квадратов (МНК)**

- **Цель:** Провести прямую через точки так, чтобы **суммарное расстояние** от точек до прямой было **минимальным**.
- **Расстояние:** Измеряется **в квадрате** (вертикальное расстояние).
    - Учитывает +/- отклонения.
    - Усиливает штраф за большие ошибки.
- **Визуализация:** Минимизация **суммарной площади квадратов** между точками и прямой.
- **Результат МНК:**  **Точные формулы** для `w` и `b`, минимизирующие сумму квадратов.
- **Вывод формул:**  Математический анализ (производные MSE по `w` и `b`).
- **Важно:**  МНК дает **точное решение** для линейной регрессии.

**Важные моменты про МНК:**

*   **Точное решение:** МНК гарантированно находит `w` и `b`, которые минимизируют MSE для линейной регрессии.
*   **Ограничение:** Применим только к линейной регрессии и ее расширениям. Не подходит для сложных моделей, таких как нейросети.
*   **Использование в Python:** Библиотеки Python (scikit-learn) используют МНК для обучения линейной регрессии автоматически.

#### **Формула МНК**
**Формула для наклона прямой (`w`):**

$$w = \frac{\sum_{i=1}^{n} [(x_i - \bar{x}) * (y_i - \bar{y})]}{\sum_{i=1}^{n} [(x_i - \bar{x})^2]}$$

**Формула для смещения (`b`):**

$$b = \bar{y} - w * \bar{x}$$

Где:

*   `xᵢ` и `yᵢ` - это значения признака (площади) и целевой переменной (цены) для каждого дома `i` в наших данных.
*   `x̄` (читается как "икс с чертой") - это **среднее значение всех признаков `xᵢ`** (средняя площадь всех домов).  Вычисляется как сумма всех `xᵢ`, деленная на количество домов.
*   `ȳ` (читается как "игрек с чертой") - это **среднее значение всех целевых переменных `yᵢ`** (средняя цена всех домов). Вычисляется как сумма всех `yᵢ`, деленная на количество домов.
*   `Σ` (сигма) - знак суммы.  В первой формуле суммирование идет по всем домам `i`.

**Формула для `w` (наклон):**

- **Числитель (ковариация):** $\boxed{\Sigma [(xᵢ - x̄) * (yᵢ - ȳ)]}$
    - Показывает, как `x` и `y` меняются **вместе**.
    - Положительный числитель: `x` и `y` растут/падают в одну сторону (положительный наклон).
    - Отрицательный числитель: `x` и `y` меняются в разные стороны (отрицательный наклон).
- **Знаменатель (дисперсия `x`):** `Σ [(xᵢ - x̄)²]`
    - Показывает **разброс** значений `x`.

**Формула для `b` (смещение):**

- $b = \boxed{ȳ - w * x̄}$
    - Линия регрессии проходит через **центр данных** `(x̄, ȳ)`.
    - Зная наклон `w` и средние `x̄`, `ȳ`, легко найти `b`.

**Главное:**

- Формулы `w` и `b` **минимизируют MSE**.
- МНК находит **лучшие** параметры линейной регрессии для данных.

### **Градиентный спуск (Gradient Descent)**

Градиентный спуск – это **алгоритм оптимизации** для **поиска минимума функции ошибки (MSE)**, то есть, оптимальных параметров `w` и `b`.

**Аналогия:**  Представьте, что вы на вершине горы в тумане и хотите спуститься в низину (минимум MSE). Вы не видите всей горы, но чувствуете уклон.  Интуитивно, вы будете двигаться в сторону **наиболее крутого спуска**.

**Шаги градиентного спуска:**

1.  **Начальная точка:** Задаем **случайные значения** `w` и `b` (как будто случайно оказались на горе).
2.  **Вычисление градиента:** Вычисляем **градиент** функции ошибки (MSE) в текущей точке. **Градиент** – это вектор, указывающий направление **наиболее крутого подъема**.
3.  **Шаг в противоположном направлении:** Делаем **шаг в сторону, противоположную градиенту** (в направлении спуска).  Изменяем `w` и `b` в сторону уменьшения MSE.  **Скорость обучения (learning rate)** определяет размер шага.
4.  **Повторение:** Повторяем шаги 2 и 3 **итеративно**, пока не достигнем **минимума MSE** (или достаточно близко).

**Важные моменты:**

*   **Итеративный метод:**  Минимум ищется **постепенно, шаг за шагом**, в отличие от прямого МНК.
*   **Скорость обучения (learning rate):**  Ключевой параметр.
    *   **Слишком мала:**  Медленный спуск, можно застрять в "пологом месте".
    *   **Слишком велика:**  Можно "перепрыгнуть" минимум и колебаться вокруг него.
*   **Локальные и глобальные минимумы:**  MSE может иметь несколько "низин". Градиентный спуск может "застрять" в **локальном минимуме**.  Но для **линейной регрессии функция MSE выпуклая**, поэтому локальных минимумов, кроме глобального, нет.

**Градиент (формально):** Вектор **частных производных** функции MSE по параметрам `w` и `b`.


Чтобы получить формулы градиентного спуска, нам сначала нужно вычислить **градиент функции ошибки (СКО)** по параметрам $w$ и $b$.  Как мы уже говорили, градиент состоит из **частных производных** СКО по каждому параметру.

Формула Среднеквадратичной ошибки (СКО):

$$MSE = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$$

где $\hat{y}_i = wx_i + b$.

1. Частная производная СКО по параметру $w$ (наклону):

$$\frac{\partial MSE}{\partial w} = -\frac{2}{N} \sum_{i=1}^{N} x_i (y_i - \hat{y}_i)$$

2. Частная производная СКО по параметру $b$ (смещению):

$$\frac{\partial MSE}{\partial b} = -\frac{2}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)$$

Разберем эти формулы по частям:

- $\frac{\partial MSE}{\partial w}$ и $\frac{\partial MSE}{\partial b}$ - это обозначения частных производных СКО по $w$ и $b$ соответственно.  Они показывают, как быстро меняется значение СКО при небольшом изменении $w$ или $b$.
- $-\frac{2}{N}$ - это константный множитель, который появляется при дифференцировании функции СКО.  Направление градиента он не меняет, но влияет на величину шага.  - знак важен, так как мы хотим двигаться в направлении *уменьшения* СКО, а градиент указывает направление *возрастания*.
- $(y_i - \hat{y}_i)$ - это ошибка предсказания для i-го объекта (разница между фактической и предсказанной ценой).
- $x_i$ в формуле для $\frac{\partial MSE}{\partial w}$ - это значение признака (площади) для i-го объекта.  Он "взвешивает" ошибку $(y_i - \hat{y}_i)$ в зависимости от значения признака $x_i$.  Чем больше $x_i$, тем больше вклад ошибки этого объекта в градиент по $w$.
- В формуле для $\frac{\partial MSE}{\partial b}$ нет $x_i$.  Здесь каждая ошибка $(y_i - \hat{y}_i)$ вносит одинаковый вклад в градиент по $b$.

Формулы обновления параметров в градиентном спуске выглядят так:

$$w_{new} = w_{old} - \alpha \frac{\partial MSE}{\partial w}$$

$$b_{new} = b_{old} - \alpha \frac{\partial MSE}{\partial b}$$

Где:

- $w_{new}$ и $b_{new}$ - это новые значения параметров $w$ и $b$, которые мы вычисляем на текущем шаге градиентного спуска.
- $w_{old}$ и $b_{old}$ - это старые значения параметров, которые были на предыдущем шаге (или случайные начальные значения на первом шаге).
- $\alpha$ (альфа) - это **скорость обучения (learning rate)**.  Это очень важный гиперпараметр, который мы задаем сами.  Он определяет размер шага, который мы делаем в направлении антиградиента на каждом шаге.

Мы берем текущие значения параметров ($w_{old}$, $b_{old}$), вычисляем градиент СКО в этой точке ($\frac{\partial MSE}{\partial w}$, $\frac{\partial MSE}{\partial b}$), умножаем градиент на скорость обучения $\alpha$ (чтобы задать размер шага), и вычитаем полученное значение из старых параметров.  **Вычитаем, потому что мы хотим двигаться в направлении *уменьшения* СКО, то есть против градиента.**

Реализация линейной регрессии с помощью градиентного спуска на Python

````python
# Создаём синтетичесие данные
import numpy as np
import matplotlib.pyplot as plt

# Истинные параметры прямой (для генерации данных)
true_w = 2
true_b = 1
num_samples = 5

x = np.linspace(0, 10, num_samples)
y = true_w * x + true_b * np.random.normal(0, 2, num_samples)

plt.scatter(x, y)
plt.xlabel('Признак (x)')
plt.ylabel('Целевая переменная (y)')
plt.title('Синтетические данные для линейной регрессии')
plt.grid()
plt.show()
````

````python
def calculate_mse(y_true, y_predicted):
    squared_errors =(y_true - y_predicted) ** 2
    mse = np.mean(squared_errors)
    return mse
````

````python
initial_w = 0
initial_b = 0

y_initital_predicted = initial_w * x + initial_b

initial_mse = calculate_mse(y, y_initital_predicted)
print(f'Начальные параметры: w = {initial_w}, b = {initial_b}')
print(f'Начальное значение MSE: {initial_mse:.2f}' if initial_mse is not None else 'Не удалось вычислить MSE')

plt.scatter(x, y)
plt.plot(x, y_initital_predicted, color='red')
plt.xlabel("Признак (x)")
plt.ylabel("Целевая переменная (y)")
plt.title("Начальное приближение линейной регрессии (MSE = {:.2f})".format(initial_mse) if initial_mse is not None else "Начальное приближение линейной регрессии (MSE не вычислена)")
plt.grid(True)
plt.show()
````

````python
def calculate_gradient_w(x, y_true, y_predicted):
    errors = y_true - y_predicted
    gradient_w = -2 * np.mean(x*errors)
    return gradient_w

def calculate_gradient_b(y_true, y_predicted):
    errors = y_true - y_predicted
    gradient_b = -2 * np.mean(errors)
    return gradient_b
````

````python
gradient_w_initial = calculate_gradient_w(x, y, y_initital_predicted)
gradient_b_initial = calculate_gradient_b(y, y_initital_predicted)

print("Начальные градиенты:")
print(f"  Градиент по w (наклону): {gradient_w_initial:.2f}")
print(f"  Градиент по b (смещению): {gradient_b_initial:.2f}")
````

Как интерпретировать значения градиентов?

gradient_w_initial = -295.34: Отрицательный градиент по w означает, что если мы немного увеличим w, то MSE уменьшится (мы двигаемся в направлении антиградиента). Значение -295.34 говорит о том, что изменение w довольно сильно влияет на MSE.

gradient_b_initial = -14.48: Отрицательный градиент по b означает, что если мы немного увеличим b, то MSE тоже уменьшится. Влияние b на MSE немного меньше, чем влияние w (градиент по b меньше по модулю).

**алгоритм градиентного спуска (Gradient Descent).**

1.  **Инициализация параметров:** Задаем начальные значения `w` и `b` (например, случайные или нули, как мы делали).
2.  **Цикл итераций:** Повторяем несколько раз (или пока не достигнем сходимости):
    *   Вычисляем **предсказанные значения** `y_predicted` для текущих `w` и `b`.
    *   Вычисляем **функцию ошибки** (MSE) для `y_predicted` и `y_true`.
    *   Вычисляем **градиенты** СКО по `w` и `b`.
    *   **Обновляем параметры** `w` и `b`, делая шаг в направлении антиградиента:
        ```
        w = w - learning_rate * gradient_w
        b = b - learning_rate * gradient_b
        ```
3.  **Результат:**  После завершения цикла получаем "оптимальные" значения `w` и `b`, которые минимизируют функцию ошибки на наших данных.


````python
learning_rate = 0.01
num_iterations = 100

current_w = 0
current_b = 0

mse_history, w_history, b_history = [], [], []

for iteration in range(num_iterations):
    y_predicted = current_w * x + current_b
    
    mse = calculate_mse(y, y_predicted)
    mse_history.append(mse)
    
    gradient_w = calculate_gradient_w(x, y, y_predicted)
    gradient_b = calculate_gradient_b(y, y_predicted)
    
    current_w = current_w - learning_rate * gradient_w
    current_b = current_b - learning_rate * gradient_b
    
    w_history.append(current_w)
    b_history.append(current_b)
    
    if (iteration + 1) % 10 == 0:
        print(f"Итерация {iteration + 1}/{num_iterations}: MSE = {mse:.2f}, w = {current_w:.2f}, b = {current_b:.2f}")
    
print("\nОбучение завершено!")
print(f"Финальные параметры: w = {current_w:.2f}, b = {current_b:.2f}")
print(f"Финальное значение MSE: {mse:.2f}")
````

Этот код выполняет градиентный спуск для обучения модели линейной регрессии.

**Основные шаги кода:**

1. **Инициализация:**
    - Задаются гиперпараметры: `learning_rate` (скорость обучения) и `num_iterations` (количество итераций).
    - Инициализируются параметры модели: `current_w` и `current_b` (начальные значения веса и смещения).
    - Создаются списки для хранения истории MSE и параметров в процессе обучения.

2. **Итерационный процесс обучения (цикл `for`):**
    - В каждой итерации:
        - Вычисляются предсказания модели (`y_predicted`) для текущих параметров.
        - Вычисляется значение функции ошибки MSE (`mse`).
        - Вычисляются градиенты MSE по `w` (`gradient_w`) и `b` (`gradient_b`).
        - Обновляются параметры `current_w` и `current_b` в направлении, противоположном градиенту, с учетом скорости обучения.
        - Сохраняется история MSE и параметров.
        - Периодически (каждые 10 итераций) выводится информация о прогрессе обучения (номер итерации, MSE, значения `w` и `b`).

3. **Завершение обучения:**
    - После завершения цикла выводится сообщение "Обучение завершено!".
    - Выводятся финальные значения параметров `w` и `b`, а также финальное значение MSE.

**Ключевая идея:** Код итеративно подстраивает параметры линейной регрессии (`w` и `b`), двигаясь в направлении, которое минимизирует среднеквадратичную ошибку (MSE) на обучающих данных.

````python
# --- Визуализация изменения MSE в процессе обучения ---
plt.figure(figsize=(8, 6)) # Задаем размер графика
plt.plot(range(1, num_iterations + 1), mse_history, marker='o', linestyle='-') # Строим график: x - номера итераций, y - mse_history
plt.xlabel("Номер итерации")
plt.ylabel("MSE")
plt.title("Динамика изменения MSE в процессе градиентного спуска")
plt.grid(True)
plt.show()
````

#### **Множественная линейная регрессия**

**Отличие от простой линейной регрессии:**

*   **Простая:** 1 признак, прямая на плоскости (2D).
*   **Множественная (2 признака):** 2 признака, плоскость в пространстве (3D).
*   **Множественная (много признаков):** Много признаков, гиперплоскость в многомерном пространстве.

**Уравнение:**

$$y = w₁x₁ + w₂x₂ + ... + wₙxₙ + b$$

*   `y` - целевая переменная.
*   `x₁, x₂, ..., xₙ` - признаки.
*   `w₁, w₂, ..., wₙ` - веса признаков (влияние каждого признака на `y` при фикс. остальных).
*   `b` - смещение.

**Задача:** Найти веса `w₁, w₂, ..., wₙ` и смещение `b`, минимизирующие MSE для гиперплоскости, наилучшим образом описывающей зависимость `y` от `x₁, x₂, ..., xₙ`.

#### **Математика множественной линейной регрессии**

**1. Векторная и матричная запись данных**

Представьте, что у нас есть данные о нескольких объектах, и для каждого объекта есть несколько признаков и целевая переменная. Мы можем представить эти данные в виде **матрицы признаков `X`** и **вектора целевых переменных `y`**.

*   **Матрица признаков `X` (Feature Matrix):**

    Это будет **матрица размера (m x n)**, где:
    *   **m** - количество объектов (количество **строк** в матрице, или количество **объектов** в данных).
    *   **n** - количество признаков (количество **столбцов** в матрице, или количество **фичей**).

    Каждая **строка** матрицы `X` будет соответствовать **одному объекту**, и содержать значения всех его признаков.
    Каждый **столбец** матрицы `X` будет соответствовать **одному признаку**, и содержать значения этого признака для всех объектов.

    **Пример:** Если у нас 3 объекта и 2 признака, матрица `X` может выглядеть так:

    ```
    X =  [ [признак1_объекта_1, признак2_объекта_1],
           [признак1_объекта_2, признак2_объекта_2],
           [признак1_объекта_2, признак2_объекта_2],
           [признак1_объекта_3, признак2_объекта_3] ]
    ```

*   **Вектор целевых переменных `y` (Target Vector):**

    Это будет **вектор-столбец размера (m x 1)**, где:
    *   **m** - количество объектов (как и в матрице `X`).

    Каждый элемент вектора `y` будет соответствовать **значению целевой переменной соответствующего объекта** (той же строке в матрице `X`).

    **Пример (для тех же 3 объектов):**

    ```
    y =  [ [значение_целевой_переменной_объекта_1],
           [значение_целевой_переменной_объекта_2],
           [значение_целевой_переменной_объекта_3] ]
    ```

*   **Вектор весов `w` (Weight Vector):**

    Это будет **вектор-столбец размера (n x 1)**, где:
    *   **n** - количество признаков (как и количество столбцов в `X`).

    Каждый элемент вектора `w` будет соответствовать **весу (коэффициенту) для соответствующего признака**.

    **Пример (для 2 признаков):**

    ```
    w =  [ [w₁],  // вес для признака 1
           [w₂] ] // вес для признака 2
    ```

*   **Смещение `b` (Bias):**

    Это будет **скаляр** (одиночное число).

**2. Уравнение множественной линейной регрессии в матричной форме**

Теперь мы можем записать уравнение множественной линейной регрессии **очень компактно и красиво** в матричной форме:

`ŷ = Xw + b`

Где:

*   `ŷ` (читается как "игрек с крышкой") - это **вектор предсказанных значений** (вектор-столбец размера (m x 1)). Это то, что мы получаем на выходе нашей модели.
*   `X` - матрица признаков (m x n).
*   `w` - вектор весов (n x 1).
*   `b` - смещение (скаляр).

**Операция `Xw` означает матричное умножение матрицы `X` на вектор `w`.** Результатом будет вектор-столбец размера (m x 1). Затем к каждому элементу этого вектора прибавляется скаляр `b`.

**Если расписать это уравнение покомпонентно, для каждого объекта `i` (от 1 до m), мы получим:**

`ŷᵢ = w₁xᵢ₁ + w₂xᵢ₂ + ... + wₙxᵢₙ + b`

где:

*   `ŷᵢ` - предсказанное значение целевой переменной i-го объекта.
*   `xᵢ₁, xᵢ₂, ..., xᵢₙ` - значения признаков для i-го объекта (i-я строка матрицы `X`).
*   `w₁, w₂, ..., wₙ` - веса признаков (элементы вектора `w`).
*   `b` - смещение.

**Это в точности то же самое уравнение, которое мы писали раньше, но теперь в компактной матричной форме!**

**3. Функция ошибки (MSE) и градиентный спуск в матричной форме**

Функция ошибки **Среднеквадратичная ошибка (MSE)** остается той же самой по сути: мы все так же хотим минимизировать средний квадрат разницы между фактическими и предсказанными значениями целевой переменной. В матричной форме MSE тоже можно записать, но пока не будем усложнять. Главное, что **принцип минимизации MSE остается прежним.**

Алгоритм **градиентного спуска** тоже остается тем же самым: мы итеративно обновляем параметры (теперь уже вектор весов `w` и смещение `b`), двигаясь в направлении антиградиента функции ошибки. Формулы для градиентов в матричной форме тоже есть, но они немного сложнее, и мы пока не будем их выводить подробно. Важно понимать, что **принцип градиентного спуска не меняется, меняется только запись данных и параметров - теперь мы используем векторы и матрицы.**

**Зачем нам вообще нужна эта матричная запись?**

*   **Компактность и наглядность:** Матричные формулы выглядят гораздо короче и понятнее, чем длинные суммы.
*   **Эффективность вычислений:** Операции с матрицами и векторами очень хорошо оптимизированы в библиотеках численных вычислений, таких как NumPy. Матричное умножение выполняется гораздо быстрее, чем поэлементные вычисления в циклах, особенно для больших объемов данных. **Использование матричной формы позволяет нам писать более быстрый и эффективный код для машинного обучения!**

**Как тебе матричная запись множественной линейной регрессии? Понятно, как матрица `X`, вектор `w` и смещение `b` связаны с уравнением и предсказаниями?** Или нужно что-то уточнить? Не стесняйся спрашивать! 😉 Дальше мы посмотрим, как использовать `scikit-learn` для множественной линейной регрессии, и как это все работает на практике! 🚀

#### **Регуляризация (Regularization).**

**Регуляризация** - это группа методов, которые используются для **борьбы с переобучением (overfitting)** моделей машинного обучения.  Переобучение - это одна из главных проблем в ML, и регуляризация - один из самых эффективных способов ее решения.

**Вспомним, что такое переобучение:**

**Переобучение** - это ситуация, когда модель **слишком хорошо "запоминает" обучающие данные**, подстраиваясь под каждую деталь и даже под шум в этих данных.  В результате, такая модель **отлично работает на обучающих данных**, показывая очень высокую точность, но **плохо обобщается на новые, unseen данные**, то есть, плохо работает на тестовых данных или в реальной жизни.

**В линейных моделях (и не только) переобучение может проявляться в том, что модель получает слишком большие по модулю веса (коэффициенты) `w`.**  Большие веса делают модель **слишком чувствительной к малейшим изменениям во входных данных**, что и приводит к переобучению.

**Регуляризация - это методы, которые "штрафуют" модель за слишком большие веса, заставляя ее выбирать более "простые" и "гладкие" решения, которые лучше обобщаются на новые данные.**

**Существует два основных типа регуляризации для линейных моделей:**

1.  **L1-регуляризация (Lasso Regularization).**
2.  **L2-регуляризация (Ridge Regularization).**

Иногда еще говорят про **Elastic Net Regularization**, которая является комбинацией L1 и L2.

**1. L2-регуляризация (Ridge Regularization)**

*   **Идея:**  Добавить к функции ошибки (MSE) **штраф за большие значения весов `w`**, а именно, **сумму квадратов всех весов**.

*   **Новая функция ошибки (с L2-регуляризацией):**

    $MSE_Regularized = MSE + λ * Σ(wᵢ²)$

    Где:
    *   `MSE` - обычная Среднеквадратичная ошибка.
    *   `Σ(wᵢ²)` - сумма квадратов всех весов модели (от `w₁` до `wₙ`).
    *   `λ` (лямбда) - **коэффициент регуляризации (regularization strength)**.  Это **гиперпараметр**, который мы задаем сами.  Он регулирует **силу регуляризации**.
        *   **Чем больше `λ`**, тем сильнее штраф за большие веса, и тем сильнее регуляризация.  Модель будет стремиться к **меньшим по модулю весам**.
        *   **Чем меньше `λ`**, тем слабее регуляризация.  При `λ = 0` регуляризация вообще отсутствует, и мы получаем обычную линейную регрессию без регуляризации.

*   **Эффект L2-регуляризации:**  L2-регуляризация **"стягивает" веса к нулю**, но обычно **не обнуляет их полностью**.  Она делает веса **меньше по величине**, но **все признаки остаются в модели** (хотя вклад "менее важных" признаков уменьшается).  L2-регуляризация помогает **уменьшить дисперсию (variance) модели**, и, таким образом, **бороться с переобучением**.

**2. L1-регуляризация (Lasso Regularization)**

*   **Идея:**  Добавить к функции ошибки (MSE) **штраф за большие значения весов `w`**, но на этот раз, в виде **суммы модулей всех весов**.

*   **Новая функция ошибки (с L1-регуляризацией):**

    `MSE_regularized = MSE + λ * Σ(|wᵢ|)`

    Где:
    *   `MSE` и `λ` - то же самое, что и в L2-регуляризации.
    *   `Σ(|wᵢ|)` - сумма модулей всех весов модели (от `w₁` до `wₙ`).

*   **Эффект L1-регуляризации:**  L1-регуляризация, в отличие от L2, может **обнулять некоторые веса полностью**.  Это означает, что **L1-регуляризация автоматически выполняет отбор признаков (feature selection)!**  "Неважные" признаки, которые мало влияют на целевую переменную, могут получить веса, равные нулю, и, таким образом, быть исключены из модели.  L1-регуляризация помогает не только **уменьшить дисперсию**, но и сделать модель **более интерпретируемой** (за счет отбора признаков).

**В чем разница между L1 и L2 регуляризацией "на пальцах"?**

*   **L2 (Ridge):**  Мягкая регуляризация, "стягивает" веса к нулю, но не обнуляет их.  Помогает уменьшить влияние всех признаков, особенно "шумных" и коррелированных.  Хорошо подходит, когда нужно **улучшить обобщающую способность модели**, не теряя при этом информацию от всех признаков.
*   **L1 (Lasso):**  Более жесткая регуляризация, может обнулять веса.  Выполняет **отбор признаков**.  Хорошо подходит, когда нужно **уменьшить переобучение и сделать модель более разреженной и интерпретируемой**, выделив только самые важные признаки.

**Выбор между L1 и L2 регуляризацией (и коэффициентом регуляризации `λ`) - это задача **настройки гиперпараметров**, которую обычно решают с помощью **кросс-валидации** (мы до этого еще дойдем).**

#### **Как применять L1 и L2 регуляризацию на практике с помощью `scikit-learn`.**

**`scikit-learn` предоставляет готовые классы для линейной регрессии с L1 и L2 регуляризацией:**

*   **`Ridge`** - для **L2-регуляризации (Гребневая регрессия)**.
*   **`Lasso`** - для **L1-регуляризации (Лассо регрессия)**.

Оба класса находятся в том же модуле `sklearn.linear_model`, где и `LinearRegression`.

**1. L2-регуляризация (Ridge Regression) в `scikit-learn`.**

````python
from sklearn.linear_model import Ridge # Импорт класса Ridge

# --- L2-регуляризация (Ridge) ---
ridge_model = Ridge(alpha=1.0) # Создаем объект Ridge-модели. alpha - коэффициент регуляризации (λ). По умолчанию alpha=1.0
ridge_model.fit(X, y) # Обучаем Ridge-модель на тех же данных X, y, что и LinearRegression

# Получаем параметры Ridge-модели
ridge_w = ridge_model.coef_
ridge_b = ridge_model.intercept_

print("Параметры Ridge-модели (L2-регуляризация, alpha=1.0):")
print(f"  w (веса признаков): {ridge_w}")
print(f"  b (смещение): {ridge_b:.2f}")

# Сравним с параметрами обычной LinearRegression (без регуляризации)
print("\nПараметры LinearRegression (без регуляризации):")
print(f"  w (веса признаков): {sklearn_w}") # sklearn_w и sklearn_b мы получили на предыдущем шаге
print(f"  b (смещение): {sklearn_b:.2f}")
````

**2. L1-регуляризация (Lasso Regression) в `scikit-learn`.**

````python
from sklearn.linear_model import Lasso # Импорт класса Lasso

# --- L1-регуляризация (Lasso) ---
lasso_model = Lasso(alpha=0.1) # Создаем объект Lasso-модели. alpha - коэффициент регуляризации (λ).  Для Lasso обычно используют меньшие alpha, чем для Ridge.
lasso_model.fit(X, y) # Обучаем Lasso-модель

# Получаем параметры Lasso-модели
lasso_w = lasso_model.coef_
lasso_b = lasso_model.intercept_

print("Параметры Lasso-модели (L1-регуляризация, alpha=0.1):")
print(f"  w (веса признаков): {lasso_w}")
print(f"  b (смещение): {lasso_b:.2f}")

# Сравним с параметрами LinearRegression и Ridge
print("\nПараметры Ridge-модели (L2-регуляризация, alpha=1.0):")
print(f"  w (веса признаков): {ridge_w}")
print(f"  b (смещение): {ridge_b:.2f}")
print("\nПараметры LinearRegression (без регуляризации):")
print(f"  w (веса признаков): {sklearn_w}")
print(f"  b (смещение): {sklearn_b:.2f}")
````

**Что делает этот код:**

*   Импортирует класс `Lasso` из `sklearn.linear_model`.
*   Создает объект `lasso_model = Lasso(alpha=0.1)`.  Здесь мы тоже задаем коэффициент регуляризации `alpha`, но на этот раз выбрали `alpha=0.1` - немного меньше, чем для `Ridge`.  Для Lasso регуляризация часто бывает более "сильной", чем для Ridge, поэтому для Lasso часто используют меньшие значения `alpha`.
*   Обучает `lasso_model.fit(X, y)`.
*   Получает параметры `lasso_w` и `lasso_b`.
*   Выводит параметры `Lasso`-модели, `Ridge`-модели и `LinearRegression`-модели для сравнения.

**Что ты должен увидеть в выводе:**

*   Параметры `Lasso`-модели, `Ridge`-модели и `LinearRegression`-модели.
*   **Сравни значения весов `w` для `Lasso`, `Ridge` и `LinearRegression`.**  Ты должен заметить, что **веса `lasso_w` еще меньше по модулю, чем `ridge_w` и `sklearn_w`.  А в некоторых случаях, один из весов `lasso_w` может стать *равным нулю* или очень близким к нулю!**  Например, `lasso_w` может быть `[1.8...  0.00]`.  В этом и проявляется эффект **отбора признаков** L1-регуляризации!  Lasso может "обнулять" веса "неважных" признаков, исключая их из модели.

**3. Коэффициент регуляризации `alpha`.**

**Параметр `alpha` (коэффициент регуляризации `λ`) в `Ridge` и `Lasso` - это **очень важный гиперпараметр!**  Он контролирует **силу регуляризации**.

*   **Чем больше `alpha`**, тем **сильнее регуляризация**.  Веса будут "стягиваться" к нулю сильнее.  Модель будет становиться **проще**, менее склонной к переобучению, но может немного **увеличиться смещение (bias)**.
*   **Чем меньше `alpha`**, тем **слабее регуляризация**.  При `alpha = 0` регуляризация вообще отсутствует, и `Ridge` и `Lasso` превращаются в обычную `LinearRegression`.  Модель будет становиться **сложнее**, более склонной к переобучению, но может **уменьшиться смещение (bias)**.

#### **Вспоминаем Bias-Variance Tradeoff.**

Мы уже говорили о том, что **Bias-Variance Tradeoff** - это компромисс между двумя видами ошибок модели:

*   **Смещение (Bias):** Ошибка из-за **упрощений, заложенных в модель**.  Модель с высоким смещением делает слишком **грубые** предположения о данных, и поэтому может **недообучиться (underfitting)**, то есть, плохо описывать и обучающие, и новые данные.
*   **Разброс (Variance):** Ошибка из-за **чувствительности модели к колебаниям в обучающих данных**.  Модель с высокой дисперсией **слишком сильно подстраивается под обучающие данные**, включая шум, и поэтому может **переобучиться (overfitting)**, то есть, отлично работать на обучающих данных, но плохо на новых.

**Цель - найти баланс между смещением и разбросом, чтобы минимизировать общую ошибку обобщения (generalization error) модели на новых данных.**

**Как регуляризация влияет на Разброс (Variance).**

**Регуляризация, особенно L1 и L2, в первую очередь направлена на уменьшение Разброса (Variance).**

**Почему?  За счет чего?**

*   **Штраф за большие веса:**  Регуляризация добавляет к функции ошибки **штраф за большие значения весов `w`**.  Это "заставляет" модель выбирать решения с **меньшими по модулю весами**.
*   **Упрощение модели:**  Модели с меньшими весами становятся **проще и "глаже"**.  Они **менее чувствительны к "мелким деталям" и шуму в обучающих данных.**  Именно высокая чувствительность к шуму и является основной причиной высокой дисперсии и переобучения.
*   **Уменьшение сложности:**  Регуляризация как бы **ограничивает "сложность" модели**, не позволяя ей "излишне изгибаться" и подстраиваться под каждую точку обучающих данных.  Она заставляет модель искать более **общие закономерности** в данных, а не "запоминать" конкретные примеры.

**Аналогия:**  Представь, что ты пытаешься нарисовать кривую, которая проходит через ряд точек, полученных с измерениями с шумом.

*   **Модель без регуляризации (высокая дисперсия):**  Попытается провести кривую **точно через каждую точку**, включая точки шума.  Кривая получится очень извилистой, "дерганой", и будет плохо описывать общую тенденцию.  Если ты немного изменишь обучающие точки (например, добавишь новые или уберешь старые), кривая может сильно измениться - **высокая дисперсия.**
*   **Модель с регуляризацией (низкая дисперсия):**  Будет искать **более гладкую и простую кривую**, которая проходит *примерно* через точки, но не обязательно через каждую.  Кривая получится менее извилистой, более устойчивой к шуму.  При небольших изменениях в обучающих данных кривая будет меняться не так сильно - **низкая дисперсия.**

**3. Как регуляризация влияет на Веса (Weights).**

И L1, и L2 регуляризация уменьшают дисперсию за счет **уменьшения величины весов `w`**:

*   **L2-регуляризация (Ridge):**  **Стягивает веса к нулю**, но обычно **не обнуляет их полностью**.  Все веса становятся меньше по модулю.
*   **L1-регуляризация (Lasso):**  Может **обнулять некоторые веса полностью**.  Веса "неважных" признаков могут стать равными нулю, и эти признаки исключаются из модели.

**4. Bias-Variance Tradeoff и Регуляризация: Компромисс.**

**Важно понимать, что регуляризация, уменьшая дисперсию, может *немного увеличить смещение (Bias)*.**

**Почему?**

*   **Упрощение модели:**  Делая модель проще и "глаже", регуляризация может **ограничить ее способность точно описывать сложные зависимости в данных.**  Модель становится менее гибкой.
*   **Потеря деталей:**  Стремясь к "простоте", модель может **упустить некоторые важные детали** в данных, которые не являются шумом, а несут полезную информацию.

**Таким образом, регуляризация - это всегда компромисс (Tradeoff).**  Мы **уменьшаем дисперсию (чтобы бороться с переобучением)**, но при этом **немного увеличиваем смещение (что может привести к недообучению, если переборщить с регуляризацией).**

**Цель - найти "золотую середину" - оптимальный уровень регуляризации, который позволит значительно уменьшить дисперсию, не сильно увеличив при этом смещение, и в итоге минимизировать общую ошибку обобщения модели.**

**Аналогия Bias-Variance-Регуляризация (мишень для дартс):**

*   **Без регуляризации (высокая дисперсия):**  Стрелы разбросаны **широко** вокруг центра мишени (высокая дисперсия), хотя центр "скопления" стрел может быть близок к центру мишени (потенциально низкое смещение, если модель достаточно сложная).
*   **Сильная регуляризация (высокое смещение, низкая дисперсия):**  Стрелы собраны в **плотную кучу** (низкая дисперсия), но эта куча сильно **смещена от центра мишени** (высокое смещение).  Веса сильно "стянуты", модель слишком простая и неточная.
*   **Оптимальная регуляризация (баланс Bias-Variance):**  Стрелы собраны в **относительно плотную кучу**, и эта куча **центрована примерно в центре мишени** (баланс смещения и разброса).  Регуляризация подобрана правильно, модель хорошо обобщает.

**Ключевая мысль:**  Регуляризация - это инструмент для управления Bias-Variance Tradeoff.  Она помогает уменьшить дисперсию за счет "стягивания" весов, но нужно правильно подобрать уровень регуляризации, чтобы не увеличить при этом смещение слишком сильно.
